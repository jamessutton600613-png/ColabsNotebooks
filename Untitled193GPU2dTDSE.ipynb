```python
# S14.3 â€” GPU (CuPy) 2D TDSE, 4Ã—1 panels at barrier height 0.3 with injected perturbations
# Adds: (I) âŸ¨TâŸ©, âŸ¨VâŸ©, âŸ¨HâŸ© time-series; (II) transmission-vs-time plot.
# Outputs:
#   /content/GQR_bar03_perturbations.mp4
#   /content/GQR_bar03_metrics.csv
#   /content/GQR_bar03_energy.csv
#   /content/GQR_bar03_energy.png
#   /content/GQR_bar03_transmission.png

# ----------------------
# Imports & GPU fallback
# ----------------------
import sys, math, os, csv
import numpy as _np
try:
    import cupy as _cp
    _ = _cp.zeros(1)
    use_gpu = True
    xp = _cp
    print("GPU mode: CuPy is available.")
except Exception as e:
    use_gpu = False
    xp = _np
    print("CPU fallback: CuPy not available -> using NumPy.", e)

import numpy as np  # host utilities only
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter

# -----------------------------
# Simulation domain & constants
# -----------------------------
Nx, Ny = 256, 128
Lx, Ly = 70.0, 35.0
dx, dy = Lx / Nx, Ly / Ny
x = xp.linspace(-Lx/2, Lx/2 - dx, Nx)
y = xp.linspace(-Ly/2, Ly/2 - dy, Ny)
X, Y = xp.meshgrid(x, y, indexing='xy')

hbar = 1.0
m    = 1.0

# Time stepping
dt              = 0.08
n_frames        = 240
steps_per_frame = 2
print(f"Total simulated time â‰ˆ {dt*n_frames*steps_per_frame:.2f} (arb units)")

# ------------------------------
# Core potential & perturbations
# ------------------------------
barrier_center = 0.0
barrier_width  = 2.0
barrier_height = 0.3

pocket_x_start = 10.0
pocket_depth   = -15.0

V0 = xp.zeros((Ny, Nx), dtype=xp.float64)
barrier_mask = xp.abs(X - barrier_center) < (barrier_width / 2.0)
V0 = V0 + barrier_height * barrier_mask
V0 = V0 + pocket_depth * (X > pocket_x_start)

# --- Complex Absorbing Boundary (CAP) ---
cap_margin = 8.0
eta = 2.5e-4  # absorption strength
def cap_profile(coord, L, margin):
    r = xp.maximum(xp.abs(coord) - (L/2 - margin), 0.0) / margin
    return r**4
CAP = -1j * eta * (cap_profile(X, Lx, cap_margin) + cap_profile(Y, Ly, cap_margin))
V0_cap = V0 + CAP  # base + absorber
# NOTE: For energies we use only the **Hermitian** part (exclude CAP). See V*_real() below.

# --------------- Initial packet ---------------
x0, y0  = -20.0, 0.0
kx0     = 1.6
ky0     = 0.0
sigx    = 3.5
sigy    = 4.0

psi_init = xp.exp(-((X - x0)**2/(2*sigx**2) + (Y - y0)**2/(2*sigy**2))).astype(xp.complex128)
psi_init *= xp.exp(1j * (kx0 * X + ky0 * Y))
norm0 = xp.sqrt(xp.sum(xp.abs(psi_init)**2) * dx * dy)
psi_init /= norm0

# ---------------------------- Split-operator precompute ----------------------------
if use_gpu:
    kx = 2*xp.pi * _cp.fft.fftfreq(Nx, d=dx)
    ky = 2*xp.pi * _cp.fft.fftfreq(Ny, d=dy)
else:
    kx = 2*xp.pi * xp.fft.fftfreq(Nx, d=dx)
    ky = 2*xp.pi * xp.fft.fftfreq(Ny, d=dy)
KX, KY = xp.meshgrid(kx, ky, indexing='xy')

T_k = (hbar**2)*(KX**2 + KY**2) / (2*m)
expT_full = xp.exp(-1j * T_k * (dt/hbar))

fft2  = xp.fft.fft2
ifft2 = xp.fft.ifft2

def step_split_operator(psi, V):
    expV_half = xp.exp(-1j * V * (dt/(2*hbar)))
    psi *= expV_half
    psi_k = fft2(psi)
    psi_k *= expT_full
    psi = ifft2(psi_k)
    psi *= expV_half
    return psi

# -------------------- Variants (with CAP) and their Hermitian parts --------------------
# B drive envelope precompute
poc_x, poc_y = 20.0, 0.0
sigp = 6.0
ENV_B = xp.exp(-(((X - poc_x)**2 + (Y - poc_y)**2)/(2*sigp**2)))

# params
A_B, w_B = 0.8, 0.12
E0_C, w_C = 0.06, 0.10
dH_D, w_D = 0.25, 0.08

# Full (for evolution)
def V_A(t): return V0_cap
def V_B(t): return V0_cap + A_B * xp.sin(w_B * t) * ENV_B
def V_C(t): return V0_cap - (E0_C * xp.sin(w_C * t)) * X
def V_D(t):
    mod = barrier_height + dH_D * xp.sin(w_D * t)
    Vb  = mod * (xp.abs(X - barrier_center) < (barrier_width/2.0))
    return (V0_cap - barrier_height * (xp.abs(X - barrier_center) < (barrier_width/2.0))) + Vb

# Hermitian (real) part for âŸ¨VâŸ© (exclude CAP)
def V_A_real(t): return V0
def V_B_real(t): return V0 + A_B * xp.sin(w_B * t) * ENV_B
def V_C_real(t): return V0 - (E0_C * xp.sin(w_C * t)) * X
def V_D_real(t):
    mod = barrier_height + dH_D * xp.sin(w_D * t)
    Vb  = mod * (xp.abs(X - barrier_center) < (barrier_width/2.0))
    return (V0 - barrier_height * (xp.abs(X - barrier_center) < (barrier_width/2.0))) + Vb

variants = [
    ("(A) control", V_A, V_A_real),
    ("(B) pocket drive", V_B, V_B_real),
    ("(C) uniform field", V_C, V_C_real),
    ("(D) barrier modulation", V_D, V_D_real),
]

# ------------------- Prepare 4 wavefunctions -------------------
psis = [psi_init.copy(), psi_init.copy(), psi_init.copy(), psi_init.copy()]

# ------------------- Matplotlib figure (density panels) -------------------
fig, axes = plt.subplots(1, 4, figsize=(18, 4.5), sharey=True)
extent = (-Lx/2, Lx/2 - dx, -Ly/2, Ly/2 - dy)
ims = []
for ax, (name, _, __) in zip(axes, variants):
    im = ax.imshow(_np.zeros((Ny, Nx)), extent=extent, origin='lower',
                   interpolation='nearest', aspect='auto', vmin=0, vmax=0.02)
    ims.append(im)
    ax.set_title(name)
    ax.set_xlabel('x')
axes[0].set_ylabel('y')

# Overlay V=0 contour for orientation
V0_host = _cp.asnumpy(V0) if use_gpu else V0
for ax in axes:
    ax.contour(_np.linspace(extent[0], extent[1], Nx),
               _np.linspace(extent[2], extent[3], Ny),
               V0_host, levels=[0.0], colors='white', linewidths=0.6, alpha=0.6)

time_text = fig.text(0.995, 0.02, '', ha='right', va='bottom', fontsize=9)

# ------------------- Regions & helpers for metrics/energies -------------------
left_region    = X < -5.0
pocket_region  = X > pocket_x_start
left_mask_host   = _cp.asnumpy(left_region)   if use_gpu else left_region
pocket_mask_host = _cp.asnumpy(pocket_region) if use_gpu else pocket_region

# Rolling vmax stabilizer
vmax_ema = [0.02, 0.02, 0.02, 0.02]
ema_decay = 0.9

metrics = []  # [t, normA, transA, reflA, ..., normD, transD, reflD]

# Energy time-series (I)
# Each row: [t, T_A, V_A, H_A, T_B, V_B, H_B, T_C, V_C, H_C, T_D, V_D, H_D]
energy_rows = []

def panel_metrics(dens_host):
    total = float(dens_host.sum() * dx * dy)
    trans = float((dens_host[pocket_mask_host]).sum() * dx * dy)
    refl  = float((dens_host[left_mask_host]).sum() * dx * dy)
    return total, trans, refl

def kinetic_energy(psi):
    """âŸ¨TâŸ© = (Ä§Â²/2m) âˆ« (|âˆ‚x Ïˆ|Â² + |âˆ‚y Ïˆ|Â²) dx dy (central differences, periodic rolls)."""
    # central differences
    psi_x = (xp.roll(psi, -1, axis=1) - xp.roll(psi, 1, axis=1)) / (2*dx)
    psi_y = (xp.roll(psi, -1, axis=0) - xp.roll(psi, 1, axis=0)) / (2*dy)
    grad2 = (xp.abs(psi_x)**2 + xp.abs(psi_y)**2)
    return float(((hbar**2)/(2*m)) * xp.sum(grad2).real * dx * dy)

def potential_energy(psi, V_real):
    """âŸ¨VâŸ© = âˆ« Ïˆ* V_real Ïˆ dx dy (V_real is purely real; CAP excluded)."""
    dens = xp.abs(psi)**2
    return float(xp.sum(dens * V_real).real * dx * dy)

progress_marks = set(int(n_frames*p/10) for p in range(1,10))

def update(frame_idx):
    global psis, vmax_ema
    t_now = dt * steps_per_frame * frame_idx

    if frame_idx in progress_marks:
        pct = int(frame_idx * 100 / n_frames)
        print(f"{pct}% done...")

    row_m = [t_now]
    row_e = [t_now]

    for k, (name, Vfun, Vreal_fun) in enumerate(variants):
        # evolve
        for _ in range(steps_per_frame):
            psis[k] = step_split_operator(psis[k], Vfun(t_now))

        # density (host)
        dens = xp.abs(psis[k])**2
        dens_host = _cp.asnumpy(dens) if use_gpu else dens

        # metrics
        tot, trans, refl = panel_metrics(dens_host)
        row_m.extend([tot, trans, refl])

        # energies (I)
        Tk = kinetic_energy(psis[k])
        Vr = potential_energy(psis[k], Vreal_fun(t_now))
        row_e.extend([Tk, Vr, Tk+Vr])

        # display autoscale
        vmax_now = float(_np.percentile(dens_host, 99.5))
        vmax_ema[k] = ema_decay * vmax_ema[k] + (1-ema_decay) * max(vmax_now, 1e-12)
        ims[k].set_data(dens_host)
        ims[k].set_clim(0, max(vmax_ema[k], 0.02))

    metrics.append(row_m)
    energy_rows.append(row_e)
    time_text.set_text(f"t = {t_now:.2f} (arb)")
    return ims + [time_text]

anim = FuncAnimation(fig, update, frames=n_frames, interval=30, blit=False)

# ------------------- Save the movie -------------------
out_mp4 = "/content/GQR_bar03_perturbations.mp4"
try:
    writer = FFMpegWriter(fps=30, bitrate=2500)
    anim.save(out_mp4, writer=writer, dpi=140)
    print(f"[OK] Saved MP4 -> {out_mp4}")
except Exception as e:
    print("[WARN] FFmpeg failed, trying GIF via PillowWriter:", e)
    out_gif = out_mp4.replace(".mp4", ".gif")
    writer = PillowWriter(fps=20)
    anim.save(out_gif, writer=writer, dpi=140)
    print(f"[OK] Saved GIF -> {out_gif}")

plt.tight_layout()
plt.show()

# ------------------- Save metrics CSV -------------------
csv_metrics = "/content/GQR_bar03_metrics.csv"
headers_m = ["t",
             "norm_A","trans_A","refl_A",
             "norm_B","trans_B","refl_B",
             "norm_C","trans_C","refl_C",
             "norm_D","trans_D","refl_D"]
with open(csv_metrics, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(headers_m)
    w.writerows(metrics)
print(f"[OK] Saved metrics -> {csv_metrics}")

# ------------------- Save energy CSV (I) -------------------
csv_energy = "/content/GQR_bar03_energy.csv"
headers_e = ["t",
             "T_A","V_A","H_A",
             "T_B","V_B","H_B",
             "T_C","V_C","H_C",
             "T_D","V_D","H_D"]
with open(csv_energy, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(headers_e)
    w.writerows(energy_rows)
print(f"[OK] Saved energies -> {csv_energy}")

# ------------------- Plot energies (I) -------------------
# One figure: 3 stacked axes (T, V, H) with 4 curves each (Aâ€“D).
import pandas as pd
E = pd.read_csv(csv_energy)

figE, (axT, axV, axH) = plt.subplots(3, 1, figsize=(9, 9), sharex=True)
axT.plot(E["t"], E["T_A"], label="A")
axT.plot(E["t"], E["T_B"], label="B")
axT.plot(E["t"], E["T_C"], label="C")
axT.plot(E["t"], E["T_D"], label="D")
axT.set_ylabel("âŸ¨TâŸ©")
axT.legend(title="Variant", ncol=4, loc="upper right", fontsize=9)

axV.plot(E["t"], E["V_A"], label="A")
axV.plot(E["t"], E["V_B"], label="B")
axV.plot(E["t"], E["V_C"], label="C")
axV.plot(E["t"], E["V_D"], label="D")
axV.set_ylabel("âŸ¨VâŸ©")

axH.plot(E["t"], E["H_A"], label="A")
axH.plot(E["t"], E["H_B"], label="B")
axH.plot(E["t"], E["H_C"], label="C")
axH.plot(E["t"], E["H_D"], label="D")
axH.set_xlabel("t (arb)")
axH.set_ylabel("âŸ¨HâŸ©")

figE.tight_layout()
energy_png = "/content/GQR_bar03_energy.png"
figE.savefig(energy_png, dpi=160)
plt.close(figE)
print(f"[OK] Saved energy plot -> {energy_png}")

# ------------------- Plot transmission curves (II) -------------------
M = pd.read_csv(csv_metrics)
figT, ax = plt.subplots(figsize=(9, 4.5))
ax.plot(M["t"], M["trans_A"], label="A")
ax.plot(M["t"], M["trans_B"], label="B")
ax.plot(M["t"], M["trans_C"], label="C")
ax.plot(M["t"], M["trans_D"], label="D")
ax.set_xlabel("t (arb)")
ax.set_ylabel("Transmission (mass in pocket)")
ax.legend(title="Variant", ncol=4, loc="best", fontsize=9)
figT.tight_layout()
trans_png = "/content/GQR_bar03_transmission.png"
figT.savefig(trans_png, dpi=160)
plt.close(figT)
print(f"[OK] Saved transmission plot -> {trans_png}")
```

```python
# 4x4 additive-combo TDSE panels (A,B,C,D); upper triangle active (unique combos), lower triangle blank
# Outputs:
#   /content/GQR_bar03_4x4_combos.mp4
#   /content/GQR_bar03_4x4_transmission.csv

import sys, os, csv
import numpy as _np
try:
    import cupy as _cp
    _ = _cp.zeros(1)
    use_gpu = True
    xp = _cp
    print("GPU mode: CuPy is available.")
except Exception as e:
    use_gpu = False
    xp = _np
    print("CPU fallback: CuPy not available -> using NumPy.", e)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter

# ---------------- Domain & constants ----------------
Nx, Ny = 256, 128
Lx, Ly = 70.0, 35.0
dx, dy = Lx / Nx, Ly / Ny
x = xp.linspace(-Lx/2, Lx/2 - dx, Nx)
y = xp.linspace(-Ly/2, Ly/2 - dy, Ny)
X, Y = xp.meshgrid(x, y, indexing='xy')

hbar = 1.0
m    = 1.0

# Time stepping
dt              = 0.08
n_frames        = 240
steps_per_frame = 2
print(f"Total simulated time â‰ˆ {dt*n_frames*steps_per_frame:.2f} (arb units)")

# ---------------- Base potential ----------------
barrier_center = 0.0
barrier_width  = 2.0
barrier_height = 0.3

pocket_x_start = 10.0
pocket_depth   = -15.0

V0 = xp.zeros((Ny, Nx), dtype=xp.float64)
barrier_mask = xp.abs(X - barrier_center) < (barrier_width / 2.0)
V0 = V0 + barrier_height * barrier_mask
V0 = V0 + pocket_depth * (X > pocket_x_start)

# Complex Absorbing Boundary (CAP)
cap_margin = 8.0
eta = 2.5e-4
def cap_profile(coord, L, margin):
    r = xp.maximum(xp.abs(coord) - (L/2 - margin), 0.0) / margin
    return r**4
CAP = -1j * eta * (cap_profile(X, Lx, cap_margin) + cap_profile(Y, Ly, cap_margin))
V0_cap = V0 + CAP

# ---------------- Initial packet ----------------
x0, y0  = -20.0, 0.0
kx0     = 1.6
ky0     = 0.0
sigx    = 3.5
sigy    = 4.0

psi_init = xp.exp(-((X - x0)**2/(2*sigx**2) + (Y - y0)**2/(2*sigy**2))).astype(xp.complex128)
psi_init *= xp.exp(1j * (kx0 * X + ky0 * Y))
norm0 = xp.sqrt(xp.sum(xp.abs(psi_init)**2) * dx * dy)
psi_init /= norm0

# ---------------- Split-operator ----------------
if use_gpu:
    kx = 2*xp.pi * _cp.fft.fftfreq(Nx, d=dx)
    ky = 2*xp.pi * _cp.fft.fftfreq(Ny, d=dy)
else:
    kx = 2*xp.pi * xp.fft.fftfreq(Nx, d=dx)
    ky = 2*xp.pi * xp.fft.fftfreq(Ny, d=dy)
KX, KY = xp.meshgrid(kx, ky, indexing='xy')

T_k = (hbar**2)*(KX**2 + KY**2) / (2*m)
expT_full = xp.exp(-1j * T_k * (dt/hbar))

fft2  = xp.fft.fft2
ifft2 = xp.fft.ifft2

def step_split_operator(psi, V):
    expV_half = xp.exp(-1j * V * (dt/(2*hbar)))
    psi *= expV_half
    psi_k = fft2(psi); psi_k *= expT_full
    psi = ifft2(psi_k)
    psi *= expV_half
    return psi

# ---------------- Variants Aâ€“D (with CAP) ----------------
# A: control
def V_A(t): return V0_cap
def V_A_real(t): return V0

# B: Boosted pocket drive (mouth-focused, time-gated pull)
poc_x, poc_y = 12.0, 0.0
sigp = 8.0
X_gate = 0.5*(xp.tanh((X - 2.0)/1.5) + 1.0)
ENV_B_core = xp.exp(-(((X - poc_x)**2 + (Y - poc_y)**2)/(2*sigp**2)))
ENV_B = ENV_B_core * X_gate
t_arrival = (0.0 - x0) / max(kx0, 1e-9)
t0   = float(t_arrival)
tauB = 5.5
def time_gate(t): return xp.exp(-((t - t0)**2)/(2.0*tauB**2))
A_dc_B = 0.60
A_ac_B = 1.20
w_B    = 0.08
def V_B(t):
    return V0_cap - (A_dc_B + A_ac_B * time_gate(t) * xp.cos(w_B*(t - t0))) * ENV_B
def V_B_real(t):
    return V0 - (A_dc_B + A_ac_B * time_gate(t) * xp.cos(w_B*(t - t0))) * ENV_B

# C: Stronger uniform field (DC+AC tilt)
E_dc_C = 0.08
E0_C   = 0.14
w_C    = 0.10
def V_C(t):       return V0_cap - (E_dc_C + E0_C * xp.sin(w_C * t)) * X
def V_C_real(t):  return V0     - (E_dc_C + E0_C * xp.sin(w_C * t)) * X

# D: Harder modulation (higher average, smaller gate)
barrier_raise_D = 0.20
dH_D            = 0.10
w_D             = 0.08
def V_D(t):
    mod = barrier_height + barrier_raise_D + dH_D * xp.sin(w_D * t)
    Vb  = mod * barrier_mask
    return (V0_cap - barrier_height * barrier_mask) + Vb
def V_D_real(t):
    mod = barrier_height + barrier_raise_D + dH_D * xp.sin(w_D * t)
    Vb  = mod * barrier_mask
    return (V0 - barrier_height * barrier_mask) + Vb

labels = ["A","B","C","D"]
V_full  = [V_A, V_B, V_C, V_D]
V_real  = [V_A_real, V_B_real, V_C_real, V_D_real]

# Helper to additively combine two variants WITHOUT double-adding V0_cap/CAP:
# V_ij = V0_cap + (V_i - V0_cap) + (V_j - V0_cap)
def make_combo(i, j):
    Vi, Vj = V_full[i], V_full[j]
    Vri, Vrj = V_real[i], V_real[j]
    def V_ij(t):     return Vi(t) + Vj(t) - V0_cap
    def V_ij_real(t):return Vri(t)+ Vrj(t)- V0
    return V_ij, V_ij_real

# ---------------- Build 4Ã—4 grid ----------------
# We will animate only the upper triangle (i<=j). Lower triangle is blank.
fig, axes = plt.subplots(4, 4, figsize=(18, 18), sharex=True, sharey=True)
extent = (-Lx/2, Lx/2 - dx, -Ly/2, Ly/2 - dy)

# Pre-create wavefunctions and potentials per active cell
active_cells = []  # list of dicts with keys: (i,j,label, psi, Vfun, Vreal)
for i in range(4):
    for j in range(4):
        ax = axes[i,j]
        if j < i:
            # blank cell
            ax.axis('off')
            continue
        # potential for (i,j)
        if i == j:
            Vfun, Vrfun = V_full[i], V_real[i]
            name = f"{labels[i]}"
        else:
            Vfun, Vrfun = make_combo(i, j)
            name = f"{labels[i]} + {labels[j]}"
        # create psi and image
        psi = psi_init.copy()
        im = ax.imshow(_np.zeros((Ny, Nx)), extent=extent, origin='lower',
                       interpolation='nearest', aspect='auto', vmin=0, vmax=0.02)
        ax.set_title(name)
        if i == 3: ax.set_xlabel('x')
        if j == 0: ax.set_ylabel('y')
        active_cells.append({"i":i, "j":j, "label":name, "psi":psi, "im":im, "V":Vfun, "Vr":Vrfun})

# Overlay V=0 contour for reference on each active ax
V0_host = _cp.asnumpy(V0) if use_gpu else V0
xs = _np.linspace(extent[0], extent[1], Nx)
ys = _np.linspace(extent[2], extent[3], Ny)
for cell in active_cells:
    ax = axes[cell["i"], cell["j"]]
    ax.contour(xs, ys, V0_host, levels=[0.0], colors='white', linewidths=0.6, alpha=0.6)

# Rolling autoscale per cell
vmax_ema = { (c["i"],c["j"]): 0.02 for c in active_cells }
ema_decay = 0.9

time_text = fig.text(0.995, 0.01, '', ha='right', va='bottom', fontsize=10)

# Transmission metrics (in pocket region X > pocket_x_start) for each active panel
pocket_region  = X > pocket_x_start
pocket_mask_host = _cp.asnumpy(pocket_region) if use_gpu else pocket_region
def transmission_of(dens_host):
    return float((dens_host[pocket_mask_host]).sum() * dx * dy)

# CSV header
csv_rows = []  # first row will be header after we know column labels

progress_marks = set(int(n_frames*p/10) for p in range(1,10))

def update(frame_idx):
    t_now = dt * steps_per_frame * frame_idx
    if frame_idx in progress_marks:
        print(f"{int(frame_idx*100/n_frames)}% done...")

    # build a row of transmissions at this time
    row = [t_now]

    for _ in range(steps_per_frame):
        for c in active_cells:
            c["psi"] = step_split_operator(c["psi"], c["V"](t_now))

    for c in active_cells:
        dens = xp.abs(c["psi"])**2
        dens_host = _cp.asnumpy(dens) if use_gpu else dens

        # transmission
        row.append(transmission_of(dens_host))

        # display with EMA autoscale
        vmax_now = float(_np.percentile(dens_host, 99.5))
        key = (c["i"], c["j"])
        vmax_ema[key] = ema_decay * vmax_ema[key] + (1-ema_decay) * max(vmax_now, 1e-12)
        c["im"].set_data(dens_host)
        c["im"].set_clim(0, max(vmax_ema[key], 0.02))

    csv_rows.append(row)
    time_text.set_text(f"t = {t_now:.2f} (arb)")
    return [c["im"] for c in active_cells] + [time_text]

# Prepare CSV header now that we know active panel labels (in consistent order)
header = ["t"] + [c["label"] for c in active_cells]

anim = FuncAnimation(fig, update, frames=n_frames, interval=30, blit=False)

# ---------------- Save animation ----------------
out_mp4 = "/content/GQR_bar03_4x4_combos.mp4"
try:
    writer = FFMpegWriter(fps=30, bitrate=2800)
    anim.save(out_mp4, writer=writer, dpi=120)
    print(f"[OK] Saved MP4 -> {out_mp4}")
except Exception as e:
    print("[WARN] FFmpeg failed, trying GIF via PillowWriter:", e)
    out_gif = out_mp4.replace(".mp4", ".gif")
    writer = PillowWriter(fps=16)
    anim.save(out_gif, writer=writer, dpi=120)
    print(f"[OK] Saved GIF -> {out_gif}")

plt.tight_layout()
plt.show()

# ---------------- Save transmission CSV ----------------
csv_path = "/content/GQR_bar03_4x4_transmission.csv"
with open(csv_path, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(header)
    w.writerows(csv_rows)
print(f"[OK] Saved transmission table -> {csv_path}")
```

```python
# === Snapshot grid (upper triangle) at 75% runtime, bright-yellow style ===
# Saves: /content/GQR_pairs_grid_frame.png

import sys, os
import numpy as _np
try:
    import cupy as _cp
    _ = _cp.zeros(1); use_gpu=True; xp=_cp; print("GPU: CuPy")
except Exception as e:
    use_gpu=False; xp=_np; print("CPU: NumPy (CuPy unavailable)", e)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import PowerNorm

# ---------------- Domain & constants ----------------
Nx, Ny = 256, 128
Lx, Ly = 70.0, 35.0
dx, dy = Lx/Nx, Ly/Ny
x = xp.linspace(-Lx/2, Lx/2 - dx, Nx)
y = xp.linspace(-Ly/2, Ly/2 - dy, Ny)
X, Y = xp.meshgrid(x, y, indexing='xy')

hbar = 1.0; m = 1.0
dt, n_frames, steps_per_frame = 0.08, 240, 2
T_total = dt*n_frames*steps_per_frame
t_snapshot = 0.75 * T_total
n_steps_snapshot = int(0.75 * n_frames * steps_per_frame)
print(f"Snapshot at t â‰ˆ {t_snapshot:.2f} (steps={n_steps_snapshot})")

# ---------------- Base potential ----------------
barrier_center=0.0; barrier_width=2.0; barrier_height=0.3
pocket_x_start=10.0; pocket_depth=-15.0

V0 = xp.zeros((Ny, Nx), dtype=xp.float64)
barrier_mask = xp.abs(X-barrier_center) < (barrier_width/2)
V0 += barrier_height*barrier_mask
V0 += pocket_depth*(X>pocket_x_start)

# CAP
cap_margin=8.0; eta=2.5e-4
def _cap(coord,L,margin):
    r = xp.maximum(xp.abs(coord)-(L/2 - margin),0)/margin
    return r**4
CAP = -1j*eta*(_cap(X,Lx,cap_margin)+_cap(Y,Ly,cap_margin))
V0_cap = V0 + CAP  # evolution uses CAP

# ---------------- Initial packet ----------------
x0,y0=-20.0,0.0; kx0,ky0=1.6,0.0; sigx, sigy=3.5,4.0
psi_init = xp.exp(-((X-x0)**2/(2*sigx**2)+(Y-y0)**2/(2*sigy**2))).astype(xp.complex128)
psi_init *= xp.exp(1j*(kx0*X+ky0*Y))
psi_init /= xp.sqrt(xp.sum(xp.abs(psi_init)**2)*dx*dy)

# ---------------- Split-operator ----------------
if use_gpu:
    kx = 2*xp.pi*_cp.fft.fftfreq(Nx, d=dx)
    ky = 2*xp.pi*_cp.fft.fftfreq(Ny, d=dy)
else:
    kx = 2*xp.pi*xp.fft.fftfreq(Nx, d=dx)
    ky = 2*xp.pi*xp.fft.fftfreq(Ny, d=dy)
KX, KY = xp.meshgrid(kx, ky, indexing='xy')
T_k = (hbar**2)*(KX**2+KY**2)/(2*m)
expT_full = xp.exp(-1j*T_k*(dt/hbar))
fft2=xp.fft.fft2; ifft2=xp.fft.ifft2

def step(psi, V):
    eVh = xp.exp(-1j*V*(dt/(2*hbar)))
    psi *= eVh
    psi_k = fft2(psi); psi_k *= expT_full
    psi = ifft2(psi_k); psi *= eVh
    return psi

# ---------------- Factor definitions ----------------
# A: control
def V_A(t): return V0_cap

# B: Pocket drive (boosted)
poc_x, poc_y = 12.0, 0.0; sigp=8.0
X_gate = 0.5*(xp.tanh((X-2.0)/1.5)+1.0)
ENV_B = xp.exp(-(((X-poc_x)**2+(Y-poc_y)**2)/(2*sigp**2))) * X_gate
t_arrival = (0.0 - x0)/max(kx0,1e-9); t0=float(t_arrival); tauB=5.5
def time_gate(t): return xp.exp(-((t-t0)**2)/(2*tauB**2))
A_dc_B, A_ac_B, w_B = 0.60, 1.20, 0.08
def V_B(t): return V0_cap - (A_dc_B + A_ac_B*time_gate(t)*xp.cos(w_B*(t-t0)))*ENV_B

# C: Strong uniform field (DC+AC tilt)
E_dc_C, E0_C, w_C = 0.08, 0.14, 0.10
def V_C(t): return V0_cap - (E_dc_C + E0_C*xp.sin(w_C*t))*X

# D: Harder barrier modulation
barrier_raise_D, dH_D, w_D = 0.20, 0.10, 0.08
def V_D(t):
    mod = barrier_height + barrier_raise_D + dH_D*xp.sin(w_D*t)
    Vb  = mod*barrier_mask
    return (V0_cap - barrier_height*barrier_mask) + Vb

# E: Temperature-like vib/noise near barrier
wE = xp.asarray([0.05, 0.11, 0.17], dtype=xp.float64)
aE = xp.asarray([0.04, 0.03, 0.02], dtype=xp.float64)
noise_amp = 0.015
rng = np.random.RandomState(42)
noise_field = xp.asarray(rng.normal(0,1,(Ny,Nx))) * barrier_mask
def V_E(t):
    harmonic = xp.sum(aE*xp.sin(wE*t))
    n = noise_amp*xp.sin(0.9*t)*noise_field
    return V0_cap + (harmonic)*barrier_mask + n

# F: Inhibitor bump at pocket mouth
inh_x, inh_sig, inh_h = 12.0, 3.5, 0.5
ENV_F = xp.exp(-((X-inh_x)**2)/(2*inh_sig**2))*(X>8.0)
def V_F(t): return V0_cap + inh_h*ENV_F

# Register factors and choose which to include
factors = [("A",V_A),("B",V_B),("C",V_C),("D",V_D),("E",V_E),("F",V_F)]
included = ["A","B","C","D","E","F"]  # change this list as needed
name2idx = {name:i for i,(name,_) in enumerate(factors)}
idxs = [name2idx[n] for n in included]
labels = [factors[i][0] for i in idxs]
Vlist  = [factors[i][1] for i in idxs]
N = len(labels)
print(f"N factors: {N}  â†’  unique cells (upper triangle incl. diag): {N*(N+1)//2}")

# Helper: additive combo without double-counting base
def make_combo(i, j):
    Vi, Vj = Vlist[i], Vlist[j]
    def V_ij(t): return Vi(t) + Vj(t) - V0_cap
    return V_ij

# Prepare wavefunctions for the active (upper-triangle) set
cells = []
for i in range(N):
    for j in range(N):
        if j < i:  # lower triangle: skip (blank)
            continue
        Vfun = Vlist[i] if i==j else make_combo(i,j)
        cells.append({"i":i,"j":j,"label": (labels[i] if i==j else f"{labels[i]}+{labels[j]}"),
                      "psi": psi_init.copy(), "V": Vfun})

# Propagate to the snapshot time
t = 0.0
for step_idx in range(n_steps_snapshot):
    t = dt*(step_idx+1)
    for c in cells:
        c["psi"] = step(c["psi"], c["V"](t))

# Render snapshot grid
# Small panels; bright-yellow using inferno + gamma<1
panel_in = 2.0  # inches per panel (shrink as needed)
fig, axes = plt.subplots(N, N, figsize=(panel_in*N, panel_in*N), sharex=True, sharey=True)
extent = (-Lx/2, Lx/2 - dx, -Ly/2, Ly/2 - dy)

# Compute percentile vmax per cell for contrast
vmax_cache = {}
for c in cells:
    dens = xp.abs(c["psi"])**2
    dens_host = _cp.asnumpy(dens) if use_gpu else dens
    vmax_cache[(c["i"],c["j"])] = float(_np.percentile(dens_host, 99.5))

for i in range(N):
    for j in range(N):
        ax = axes[i,j] if N>1 else axes
        ax.set_xticks([]); ax.set_yticks([])
        if j < i:
            ax.axis('off'); continue
        c = next(cc for cc in cells if cc["i"]==i and cc["j"]==j)
        dens = xp.abs(c["psi"])**2
        dens_host = _cp.asnumpy(dens) if use_gpu else dens
        vmax = max(vmax_cache[(i,j)], 0.02)
        im = ax.imshow(dens_host, extent=extent, origin='lower',
                       interpolation='nearest', aspect='auto',
                       cmap='inferno',
                       norm=PowerNorm(gamma=0.75, vmin=0.0, vmax=vmax))  # âœ… fix here
        ax.set_title(c["label"], fontsize=8, pad=2)

# Optional: a single colorbar for the whole figure (small)
# --- Colorbar in empty lower-triangle area ---
# Adjust [x0, y0, width, height] if needed (0=left/bottom, 1=right/top)
cax = fig.add_axes([0.05, 0.05, 0.02, 0.3])
cbar = fig.colorbar(im, cax=cax)
cbar.ax.tick_params(labelsize=7)
cbar.set_label("Probability density", fontsize=8)


fig.suptitle(f"Snapshot at t â‰ˆ {t_snapshot:.2f} (75% of run) â€” upper triangle = unique pairs", fontsize=10)
plt.tight_layout(rect=[0,0,1,0.97])
out_png = "/content/GQR_pairs_grid_frame.png"
fig.savefig(out_png, dpi=300)
plt.close(fig)
print(f"[OK] Saved snapshot grid -> {out_png}")
```
