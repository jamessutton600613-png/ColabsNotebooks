```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import datetime

# Data points extracted from the "Analytical Report on the Food Security and Humanitarian Crisis in the Gaza Strip (2024-2025)"
# Each tuple contains: (date, calorie_estimate, source, details)
data_points = [
    # Source: Section 1.3 - "from January to February 2024, the more than 300,000 people trapped in northern Gaza were forced to survive on an average of just 245 calories per day."
    (datetime.date(2024, 2, 15), 245, 'Oxfam/IPC', 'Northern Gaza Avg\n(Jan-Feb 2024)'),

    # Source: Table 1 - Data from Fliss-Isakov et al. based on COGAT. These represent availability at the border, not consumption.
    (datetime.date(2025, 1, 31), 3076, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'),
    (datetime.date(2025, 2, 28), 1741, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'), # Corrected day to 28
    (datetime.date(2025, 3, 31), 3446, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'),
    (datetime.date(2025, 4, 30), 4580, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'),

    # Source: Section 1.3 - "A Food and Agriculture Organization (FAO) report found that between October 2023 and December 2024, the average Gazan's nutritional intake was only 72% of the minimum required for survival."
    # Calculation: 2100 kcal * 0.72 = 1512 kcal. Date represents end of period.
    (datetime.date(2024, 12, 31), 1512, 'FAO', 'Gaza Avg Consumption\n(Oct 23 - Dec 24)'),

    # Source: Section 1.3 - "By May 2025... Data simulations from the FAO indicated that the average daily energy intake had plummeted to just 1,400 calories"
    (datetime.date(2025, 5, 15), 1400, 'FAO', 'Gaza Avg Consumption\n(Simulation)'),
]

# Sort data points by date for cleaner plotting
data_points.sort(key=lambda x: x[0])

# Separate data for plotting
dates = [point[0] for point in data_points]
calories = [point[1] for point in data_points]
sources = [point[2] for point in data_points]
details = [point[3] for point in data_points]

# Recommended daily calorie intake (Sphere Standard)
recommended_calories = 2100

# Create the plot
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(16, 10))

# Plot data points with different colors for different sources
# Green for availability data, shades of red/blue for consumption data
colors = {
    'Oxfam/IPC': '#d62728', # Red for catastrophic consumption
    'FAO': '#1f77b4',       # Blue for UN agency consumption data
    'COGAT/Fliss-Isakov': '#2ca02c' # Green for availability-at-border data
}
for i in range(len(dates)):
    ax.scatter(dates[i], calories[i], color=colors[sources[i]], s=150, zorder=5, alpha=0.9, edgecolors='black')
    ax.text(dates[i], calories[i] + 120, f'{calories[i]} kcal\n({sources[i]})\n{details[i]}', fontsize=10, ha='center', va='bottom')

# Plot the recommended calorie intake line
ax.axhline(y=recommended_calories, color='grey', linestyle='--', linewidth=2)
ax.text(max(dates), recommended_calories + 70, f'Minimum Recommended Daily Intake ({recommended_calories} kcal)', color='grey', ha='right', fontsize=11)


# Format the plot
ax.set_title('Contrasting Food Availability vs. Consumption in Gaza (2024-2025)', fontsize=20, pad=20)
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('Calories (kcal per person per day)', fontsize=14)
ax.set_ylim(0, max(calories) + 800)
ax.grid(True, which='both', linestyle='--', linewidth=0.6)

# Create legend
legend_patches = [mpatches.Patch(color=color, label=source) for source, color in colors.items()]
ax.legend(handles=legend_patches, title='Data Sources & Type', loc='upper left', fontsize=11)

# Add an annotation to explain the data contrast
ax.annotate('Data shows a chasm between high caloric *availability* at the border (green)\nand catastrophic *consumption* on the ground (red, blue).',
            xy=(0.05, 0.95), xycoords='axes fraction', ha='left', va='top',
            fontsize=12, bbox=dict(boxstyle="round,pad=0.5", fc="ivory", ec="black", lw=1))


plt.xticks(rotation=30, ha='right')
plt.tight_layout()

# Save the chart to a file
plt.savefig('gaza_food_availability_vs_consumption_chart.png', dpi=300)

print("Updated chart saved as gaza_food_availability_vs_consumption_chart.png")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import datetime

# Data points extracted from the "Analytical Report on the Food Security and Humanitarian Crisis in the Gaza Strip (2024-2025)"
# Each tuple contains: (date, calorie_estimate, source, details)
data_points = [
    # Source: Section 1.3 - "from January to February 2024, the more than 300,000 people trapped in northern Gaza were forced to survive on an average of just 245 calories per day."
    (datetime.date(2024, 2, 15), 245, 'Oxfam/IPC', 'Northern Gaza Avg\n(Jan-Feb 2024)'),

    # Source: Table 1 - Data from Fliss-Isakov et al. based on COGAT. These represent availability at the border, not consumption.
    (datetime.date(2025, 1, 31), 3076, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'),
    (datetime.date(2025, 2, 28), 1741, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'), # Corrected day to 28
    (datetime.date(2025, 3, 31), 3446, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'),
    (datetime.date(2025, 4, 30), 4580, 'COGAT/Fliss-Isakov', 'Avg Availability at Border'),

    # Source: Section 1.3 - "A Food and Agriculture Organization (FAO) report found that between October 2023 and December 2024, the average Gazan's nutritional intake was only 72% of the minimum required for survival."
    # Calculation: 2100 kcal * 0.72 = 1512 kcal. Date represents end of period.
    (datetime.date(2024, 12, 31), 1512, 'FAO', 'Gaza Avg Consumption\n(Oct 23 - Dec 24)'),

    # Source: Section 1.3 - "By May 2025... Data simulations from the FAO indicated that the average daily energy intake had plummeted to just 1,400 calories"
    (datetime.date(2025, 5, 15), 1400, 'FAO', 'Gaza Avg Consumption\n(Simulation)'),
]

# Sort data points by date for cleaner plotting
data_points.sort(key=lambda x: x[0])

# Separate data for plotting
dates = [point[0] for point in data_points]
calories = [point[1] for point in data_points]
sources = [point[2] for point in data_points]
details = [point[3] for point in data_points]

# Recommended daily calorie intake (Sphere Standard)
recommended_calories = 2100

# Create the plot
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(16, 10))

# Plot data points with different colors for different sources
# Green for availability data, shades of red/blue for consumption data
colors = {
    'Oxfam/IPC': '#d62728', # Red for catastrophic consumption
    'FAO': '#1f77b4',       # Blue for UN agency consumption data
    'COGAT/Fliss-Isakov': '#2ca02c' # Green for availability-at-border data
}
for i in range(len(dates)):
    ax.scatter(dates[i], calories[i], color=colors[sources[i]], s=150, zorder=5, alpha=0.9, edgecolors='black')
    ax.text(dates[i], calories[i] + 120, f'{calories[i]} kcal\n({sources[i]})\n{details[i]}', fontsize=10, ha='center', va='bottom')

# Plot the recommended calorie intake line
ax.axhline(y=recommended_calories, color='grey', linestyle='--', linewidth=2)
ax.text(max(dates), recommended_calories + 70, f'Minimum Recommended Daily Intake ({recommended_calories} kcal)', color='grey', ha='right', fontsize=11)


# Format the plot
ax.set_title('Contrasting Food Availability vs. Consumption in Gaza (2024-2025)', fontsize=20, pad=20)
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('Calories (kcal per person per day)', fontsize=14)
ax.set_ylim(0, max(calories) + 800)
ax.grid(True, which='both', linestyle='--', linewidth=0.6)

# Create legend
legend_patches = [mpatches.Patch(color=color, label=source) for source, color in colors.items()]
ax.legend(handles=legend_patches, title='Data Sources & Type', loc='upper left', fontsize=11)

# Add an annotation to explain the data contrast
ax.annotate('Data shows a chasm between high caloric *availability* at the border (green)\nand catastrophic *consumption* on the ground (red, blue).',
            xy=(0.05, 0.95), xycoords='axes fraction', ha='left', va='top',
            fontsize=12, bbox=dict(boxstyle="round,pad=0.5", fc="ivory", ec="black", lw=1))


plt.xticks(rotation=30, ha='right')
plt.tight_layout()

# Save the chart to a file
plt.savefig('gaza_food_availability_vs_consumption_chart.png', dpi=300)

print("Updated chart saved as gaza_food_availability_vs_consumption_chart.png")
```
