```python
# S14.3 â€” 2D Tunneling Movie (CPU-only, split-operator TDSE)
# Saves: S14_3_tunneling.mp4  and shows the last frame as a plot
# Dependencies: numpy, matplotlib (and ffmpeg installed for mp4). Falls back to Pillow if no ffmpeg.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter

# -----------------------------
# Simulation parameters (knobs)
# -----------------------------
Nx, Ny = 192, 128          # grid size (reduce for faster; e.g., 128x96)
Lx, Ly = 200.0, 120.0      # spatial extent (arbitrary units)
dx, dy = Lx / Nx, Ly / Ny
x = np.linspace(-Lx/2, Lx/2 - dx, Nx)
y = np.linspace(-Ly/2, Ly/2 - dy, Ny)
X, Y = np.meshgrid(x, y, indexing='xy')   # shape (Ny, Nx)

hbar = 1.0
m    = 1.0

# time stepping
dt        = 0.1            # time step
n_frames  = 240            # total frames in the animation
steps_per_frame = 2        # sub-steps per rendered frame (increase for smoother propagation)

# -----------------------------------
# Potential: barrier + reaction pocket
# -----------------------------------
barrier_center = 0.0
barrier_width  = 4.0
barrier_height = 12.0

pocket_x_start = 20.0
pocket_depth   = -30.0

V = np.zeros((Ny, Nx), dtype=np.float64)
# Barrier as a slab |x - x0| < w/2
barrier_mask = (np.abs(X - barrier_center) < (barrier_width/2))
V[barrier_mask] += barrier_height
# Reaction pocket to the right
V[X > pocket_x_start] += pocket_depth
# Soft edge confinement to reduce wrap-around artifacts
edge_soft = 20.0
V += 0.02 * (np.tanh((np.abs(X) - (Lx/2 - edge_soft))/2) + 1.0)
V += 0.02 * (np.tanh((np.abs(Y) - (Ly/2 - edge_soft))/2) + 1.0)

# ------------------------
# Initial wavepacket (2D)
# ------------------------
x0, y0  = -60.0, 0.0   # start left of the barrier
kx0     = 1.0          # mean momentum toward +x
ky0     = 0.0
sigx    = 6.0
sigy    = 8.0

gauss = np.exp(-((X - x0)**2/(2*sigx**2) + (Y - y0)**2/(2*sigy**2)))
phase = np.exp(1j * (kx0 * X + ky0 * Y))
psi = (gauss * phase).astype(np.complex128)   # ensure complex dtype

# normalize
norm = np.sqrt(np.sum(np.abs(psi)**2) * dx * dy)
psi /= norm

# --------------------------------------------
# Split-operator Fourier method (periodic BCs)
# --------------------------------------------
kx = 2*np.pi * np.fft.fftfreq(Nx, d=dx)
ky = 2*np.pi * np.fft.fftfreq(Ny, d=dy)
KX, KY = np.meshgrid(kx, ky, indexing='xy')
T_k = (hbar**2)*(KX**2 + KY**2) / (2*m)

expV_half = np.exp(-1j * V * (dt/(2*hbar)))
expT_full = np.exp(-1j * T_k * (dt/hbar))

fft2  = np.fft.fft2
ifft2 = np.fft.ifft2

def step_split_operator(psi):
    psi *= expV_half
    psi_k = fft2(psi)
    psi_k *= expT_full
    psi = ifft2(psi_k)
    psi *= expV_half
    return psi

# -------------------
# Animation setup
# -------------------
fig, ax = plt.subplots(figsize=(8, 4.8))
im = ax.imshow(np.abs(psi)**2, extent=[x.min(), x.max(), y.min(), y.max()],
               origin='lower', interpolation='nearest', aspect='auto')
ax.set_title('2D Tunneling: |Ïˆ(x,y,t)|^2')
ax.set_xlabel('x'); ax.set_ylabel('y')
cbar = plt.colorbar(im, ax=ax); cbar.set_label('Probability density')

# Potential overlay (optional)
levels = [0.0, 5.0, 10.0, pocket_depth/2]
cs = ax.contour(X, Y, V, levels=levels, colors='white', linewidths=0.6, alpha=0.6)
ax.clabel(cs, inline=True, fontsize=7, fmt='%g')

text_t = ax.text(0.02, 0.97, '', transform=ax.transAxes, color='white',
                 fontsize=10, va='top', ha='left',
                 bbox=dict(facecolor='black', alpha=0.4, boxstyle='round,pad=0.2'))

def update(frame):
    global psi
    for _ in range(steps_per_frame):
        psi = step_split_operator(psi)
    density = np.abs(psi)**2
    im.set_data(density)
    im.set_clim(0, np.percentile(density, 99.5))
    text_t.set_text(f't = {dt*steps_per_frame*(frame+1):.1f}')
    return (im, text_t)

anim = FuncAnimation(fig, update, frames=n_frames, interval=30, blit=False)

def save_animation(anim, filename_mp4='S14_3_tunneling.mp4', fps=30):
    try:
        writer = FFMpegWriter(fps=fps, bitrate=2500)
        anim.save(filename_mp4, writer=writer, dpi=160)
        print(f"[OK] Saved MP4 -> {filename_mp4}")
    except Exception as e:
        print("[WARN] Could not use ffmpeg; falling back to GIF via Pillow:", e)
        gif_name = filename_mp4.replace('.mp4', '.gif')
        writer = PillowWriter(fps=fps)
        anim.save(gif_name, writer=writer, dpi=160)
        print(f"[OK] Saved GIF -> {gif_name}")

# Save and also show the last frame
save_animation(anim)
plt.tight_layout()
plt.show()
```

```python
# S14.3 â€” 2D Tunneling Movie (CPU-only, split-operator TDSE, H vs D side-by-side)
# Saves: S14_3_tunneling_CAP.mp4 (MP4 if ffmpeg present; else GIF fallback)
# Dependencies: numpy, matplotlib. Optional: ffmpeg system binary for mp4.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter

# -----------------------------
# Simulation parameters (knobs)
# -----------------------------
# Grid / domain (reduce Nx,Ny for speed; increase for quality)
Nx, Ny = 192, 128
Lx, Ly = 200.0, 120.0
dx, dy = Lx / Nx, Ly / Ny
x = np.linspace(-Lx/2, Lx/2 - dx, Nx, dtype=np.float32)
y = np.linspace(-Ly/2, Ly/2 - dy, Ny, dtype=np.float32)
X, Y = np.meshgrid(x, y, indexing='xy')

# Physics constants
hbar   = np.float32(1.0)
m_H    = np.float32(1.0)  # "light" mass
m_D    = np.float32(2.0)  # "heavy" mass  (lower tunneling)
dtypeC = np.complex64     # speed-up vs complex128

# Time stepping
dt              = np.float32(0.08)   # stable step for these grids
n_frames        = 360                 # total frames in the movie
steps_per_frame = 3                   # sub-steps per rendered frame

# -----------------------------------
# Potential: barrier + reaction pocket
# -----------------------------------
# Barrier near x ~ 0, thinner & a bit lower to let H transmit
barrier_center = np.float32(0.0)
barrier_width  = np.float32(3.0)     # thinner â†’ more transmission
barrier_height = np.float32(8.0)     # lower â†’ more transmission

# Pocket: deeper capture region to the right of barrier
pocket_x_start = np.float32(18.0)
pocket_depth   = np.float32(-35.0)

V = np.zeros((Ny, Nx), dtype=np.float32)

# Add barrier (crisp slab)
barrier_mask = (np.abs(X - barrier_center) < (barrier_width/2))
V[barrier_mask] += barrier_height

# Add reaction pocket region
V[X > pocket_x_start] += pocket_depth

# Slight right-hand wall to keep the pocket from being totally open
V[X > (Lx/2 - 20.0)] += 10.0

# -------------------------------
# Complex absorbing boundaries (CAP)
# -------------------------------
# Quadratic CAP ramps near edges to absorb outgoing waves
cap_thick_x = np.float32(24.0)
cap_thick_y = np.float32(18.0)
x_cap_start = (Lx/2 - cap_thick_x)
y_cap_start = (Ly/2 - cap_thick_y)
eta_cap     = np.float32(1.5e-3)   # absorb strength (tune as needed)

Wx = np.zeros_like(V)
Wy = np.zeros_like(V)

# distance from vertical edges (in +x half; mirror by abs)
dx_edge = np.maximum(0.0, np.abs(X) - x_cap_start)
Wx = (dx_edge / cap_thick_x) ** 2

# distance from horizontal edges
dy_edge = np.maximum(0.0, np.abs(Y) - y_cap_start)
Wy = (dy_edge / cap_thick_y) ** 2

CAP = eta_cap * (Wx + Wy)  # non-negative

# Effective complex potential: V_eff = V - i * CAP
# Split-operator uses exp(-i * V_eff * dt) = exp(-i V dt) * exp(- CAP dt)
# We'll fold this into expV_half below.

# ------------------------
# Initial wavepackets (2D)
# ------------------------
x0, y0  = np.float32(-60.0), np.float32(0.0)  # start left of the barrier
kx0     = np.float32(1.2)                     # momentum toward +x
ky0     = np.float32(0.0)
sigx    = np.float32(6.0)
sigy    = np.float32(8.0)

psi0_amp = np.exp(-((X - x0)**2/(2*sigx**2) + (Y - y0)**2/(2*sigy**2))).astype(dtypeC)
phase    = np.exp(1j * (kx0 * X + ky0 * Y)).astype(dtypeC)
psi0     = (psi0_amp * phase).astype(dtypeC)

def normalize(psi):
    norm = np.sqrt(np.sum(np.abs(psi)**2) * dx * dy, dtype=np.float64)
    if norm == 0:
        return psi
    return (psi / norm).astype(dtypeC)

psi_H = normalize(psi0.copy())
psi_D = normalize(psi0.copy())

# --------------------------------------------
# Split-operator Fourier method (periodic BCs)
# --------------------------------------------
# Momentum-space grids (FFT frequencies)
kx = (2*np.pi * np.fft.fftfreq(Nx, d=dx)).astype(np.float32)
ky = (2*np.pi * np.fft.fftfreq(Ny, d=dy)).astype(np.float32)
KX, KY = np.meshgrid(kx, ky, indexing='xy')
K2 = (KX**2 + KY**2).astype(np.float32)

# Kinetic-energy operators in k-space for H and D
T_k_H = (hbar**2) * K2 / (2.0 * m_H)
T_k_D = (hbar**2) * K2 / (2.0 * m_D)

# Precompute phase factors
# exp(-i V_eff dt/2) = exp(-i V dt/2) * exp(- CAP dt/2)
expV_half = np.exp((-1j * V * (dt/(2*hbar)) - CAP * (dt/(2*hbar))).astype(dtypeC))
expT_H    = np.exp((-1j * T_k_H * (dt/hbar)).astype(dtypeC))
expT_D    = np.exp((-1j * T_k_D * (dt/hbar)).astype(dtypeC))

fft2  = np.fft.fft2
ifft2 = np.fft.ifft2

def step_split_operator(psi, expT):
    # half potential (with CAP)
    psi *= expV_half
    # full kinetic
    psi_k = fft2(psi)
    psi_k *= expT
    psi = ifft2(psi_k).astype(dtypeC)
    # half potential (with CAP)
    psi *= expV_half
    return psi

# -------------------
# Animation setup
# -------------------
fig, axes = plt.subplots(1, 2, figsize=(10.5, 4.6), constrained_layout=True)
axH, axD = axes

# initial densities
densH = np.abs(psi_H)**2
densD = np.abs(psi_D)**2

# pick a robust color scale from initial
vmax0 = np.percentile(np.hstack([densH.ravel(), densD.ravel()]), 99.5)

imH = axH.imshow(densH, extent=[x.min(), x.max(), y.min(), y.max()],
                 origin='lower', interpolation='nearest', aspect='auto', vmin=0, vmax=vmax0)
imD = axD.imshow(densD, extent=[x.min(), x.max(), y.min(), y.max()],
                 origin='lower', interpolation='nearest', aspect='auto', vmin=0, vmax=vmax0)

for ax, label in zip((axH, axD), ('H (light mass)', 'D (heavy mass)')):
    ax.set_title(label)
    ax.set_xlabel('x')
axH.set_ylabel('y')

# Overlay a couple of potential contours (real V only)
levels = [0.0, 4.0, 8.0, pocket_depth/2.0]
for ax in (axH, axD):
    cs = ax.contour(X, Y, V, levels=levels, colors='white', linewidths=0.6, alpha=0.6)
    ax.clabel(cs, inline=True, fontsize=7, fmt='%g')

txt = fig.text(0.02, 0.98, '', ha='left', va='top', color='k',
               bbox=dict(facecolor='white', alpha=0.5, boxstyle='round,pad=0.2'))

def update(frame):
    global psi_H, psi_D
    # advance both packets several sub-steps
    for _ in range(steps_per_frame):
        psi_H = step_split_operator(psi_H, expT_H)
        psi_D = step_split_operator(psi_D, expT_D)

    dH = np.abs(psi_H)**2
    dD = np.abs(psi_D)**2

    # auto-contrast but stable (avoid spikes)
    vmax = np.percentile(np.hstack([dH.ravel(), dD.ravel()]), 99.5)
    imH.set_data(dH); imH.set_clim(0, vmax)
    imD.set_data(dD); imD.set_clim(0, vmax)

    t_now = float(dt) * steps_per_frame * (frame + 1)
    txt.set_text(f't = {t_now:.2f}')

    return (imH, imD, txt)

anim = FuncAnimation(fig, update, frames=n_frames, interval=30, blit=False)

# Try to save MP4 with ffmpeg, fallback to GIF if ffmpeg unavailable
def save_animation(anim, filename_mp4='S14_3_tunneling_CAP.mp4', fps=30):
    try:
        writer = FFMpegWriter(fps=fps, bitrate=2500)
        anim.save(filename_mp4, writer=writer, dpi=140)
        print(f"[OK] Saved MP4 -> {filename_mp4}")
    except Exception as e:
        print("[WARN] Could not use ffmpeg; falling back to GIF via Pillow:", e)
        gif_name = filename_mp4.replace('.mp4', '.gif')
        writer = PillowWriter(fps=fps)
        anim.save(gif_name, writer=writer, dpi=140)
        print(f"[OK] Saved GIF -> {gif_name}")

# Save and also display the final frame
save_animation(anim)
plt.show()
```

```python
# S14.3 â€” 2D Tunneling Movie (CPU-only, split-operator TDSE)
# Saves: S14_3_tunneling_H_vs_D.mp4  (falls back to GIF if ffmpeg missing)
# Shows final frame inline; prints progress logs.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter

# -----------------------------
# Simulation parameters (knobs)
# -----------------------------
np.random.seed(0)

Nx, Ny = 160, 112         # grid size (reduce for faster; e.g., 128x96)
Lx, Ly = 200.0, 120.0     # spatial extent
dx, dy = Lx / Nx, Ly / Ny
x = np.linspace(-Lx/2, Lx/2 - dx, Nx)
y = np.linspace(-Ly/2, Ly/2 - dy, Ny)
X, Y = np.meshgrid(x, y, indexing='xy')

hbar = 1.0
m_H  = 1.0                # "light" mass (H)
m_D  = 2.0                # "heavy" mass (D) ~ 2x for simple KIE demonstration

# time stepping
dt        = 0.1           # time step
n_frames  = 260           # total frames in the animation
steps_per_frame = 2       # sub-steps per rendered frame (increase for smoother propagation)
LOG_EVERY = max(1, n_frames // 20)  # print ~5% progress steps

# -----------------------------------
# Potential: barrier + reaction pocket
# -----------------------------------
barrier_center = 0.0
barrier_width  = 4.0
barrier_height = 10.0

pocket_x_start = 15.0
pocket_depth   = -25.0
pocket_width   = 40.0   # finite-width pocket (to reduce standing-wave recurrences)

V = np.zeros((Ny, Nx), dtype=np.float64)

# Barrier (rectangular slab around x=0)
barrier_mask = (np.abs(X - barrier_center) < (barrier_width/2))
V[barrier_mask] += barrier_height

# Reaction pocket: rectangular well to the right of the barrier
pocket_mask = (X > pocket_x_start) & (X < pocket_x_start + pocket_width)
V[pocket_mask] += pocket_depth

# Soft confining walls at edges (reduce wrap-around feeling)
edge_soft = 18.0
V += 0.02 * (np.tanh((np.abs(X) - (Lx/2 - edge_soft))/2) + 1.0)
V += 0.02 * (np.tanh((np.abs(Y) - (Ly/2 - edge_soft))/2) + 1.0)

# ------------------------
# Initial wavepackets
# ------------------------
x0, y0  = -60.0, 0.0   # start left of the barrier
kx0     = 1.0          # mean momentum toward +x
ky0     = 0.0
sigx    = 7.0
sigy    = 9.0

psi0 = np.exp(-((X - x0)**2/(2*sigx**2) + (Y - y0)**2/(2*sigy**2)))
phase0 = np.exp(1j * (kx0 * X + ky0 * Y))
psi_H = (psi0 * phase0).astype(np.complex128, copy=True)
psi_D = psi_H.copy()

# normalize each
def normalize(psi):
    norm = np.sqrt(np.sum(np.abs(psi)**2) * dx * dy)
    if norm == 0:
        return psi
    return psi / norm

psi_H = normalize(psi_H)
psi_D = normalize(psi_D)

# --------------------------------------------
# Split-operator Fourier method (periodic BCs)
# --------------------------------------------
kx = 2*np.pi * np.fft.fftfreq(Nx, d=dx)
ky = 2*np.pi * np.fft.fftfreq(Ny, d=dy)
KX, KY = np.meshgrid(kx, ky, indexing='xy')

# Kinetic energy operator in k-space (mass-specific)
T_k_H = (hbar**2)*(KX**2 + KY**2) / (2*m_H)
T_k_D = (hbar**2)*(KX**2 + KY**2) / (2*m_D)

# Precompute half-step potential and full-step kinetic phases
expV_half = np.exp(-1j * V * (dt/(2*hbar)))
expT_H    = np.exp(-1j * T_k_H * (dt/hbar))
expT_D    = np.exp(-1j * T_k_D * (dt/hbar))

fft2  = np.fft.fft2
ifft2 = np.fft.ifft2

def step_split_operator(psi, expT_full):
    # 1) half step potential
    psi *= expV_half
    # 2) full step kinetic in k-space
    psi_k = fft2(psi)
    psi_k *= expT_full
    psi = ifft2(psi_k)
    # 3) half step potential
    psi *= expV_half
    return psi

# -------------------
# Animation setup
# -------------------
fig, (axH, axD) = plt.subplots(1, 2, figsize=(12, 4.8), sharey=True)

imH = axH.imshow(np.abs(psi_H)**2, extent=[x.min(), x.max(), y.min(), y.max()],
                 origin='lower', interpolation='nearest', aspect='auto')
imD = axD.imshow(np.abs(psi_D)**2, extent=[x.min(), x.max(), y.min(), y.max()],
                 origin='lower', interpolation='nearest', aspect='auto')

axH.set_title('H (light mass) â€” faster tunneling')
axD.set_title('D (heavy mass) â€” slower tunneling')
for ax in (axH, axD):
    ax.set_xlabel('x')
axH.set_ylabel('y')
cbar = fig.colorbar(imH, ax=[axH, axD], shrink=0.9)
cbar.set_label('Probability density')

# Contour overlay of the potential (ensure strictly increasing levels)
levels = np.array([min(0.0, pocket_depth), pocket_depth/2.0, 0.0, 4.0, 8.0, 12.0], dtype=float)
levels = np.unique(np.sort(levels))
for ax in (axH, axD):
    cs = ax.contour(X, Y, V, levels=levels, colors='white', linewidths=0.6, alpha=0.6)
    ax.clabel(cs, inline=True, fontsize=7, fmt='%g')

textH = axH.text(0.02, 0.97, '', transform=axH.transAxes, color='white',
                 fontsize=10, va='top', ha='left',
                 bbox=dict(facecolor='black', alpha=0.4, boxstyle='round,pad=0.2'))
textD = axD.text(0.02, 0.97, '', transform=axD.transAxes, color='white',
                 fontsize=10, va='top', ha='left',
                 bbox=dict(facecolor='black', alpha=0.4, boxstyle='round,pad=0.2'))

subtitle = fig.suptitle('2D Tunneling â€” KIE Demo (H vs D)', fontsize=14)

def update(frame):
    global psi_H, psi_D
    for _ in range(steps_per_frame):
        psi_H = step_split_operator(psi_H, expT_H)
        psi_D = step_split_operator(psi_D, expT_D)
    dH = np.abs(psi_H)**2
    dD = np.abs(psi_D)**2
    imH.set_data(dH)
    imD.set_data(dD)
    # robust auto-contrast
    imH.set_clim(0, np.percentile(dH, 99.5))
    imD.set_clim(0, np.percentile(dD, 99.5))
    t_now = dt * steps_per_frame * (frame + 1)
    textH.set_text(f't = {t_now:.1f}')
    textD.set_text(f't = {t_now:.1f}')
    # progress logs
    if (frame + 1) % LOG_EVERY == 0 or frame == 0:
        pct = int(100 * (frame + 1) / n_frames)
        print(f"[S14.3] Renderingâ€¦ {pct}% ({frame+1}/{n_frames} frames)")
    return (imH, imD, textH, textD)

anim = FuncAnimation(fig, update, frames=n_frames, interval=30, blit=False)

def save_animation(anim, filename_mp4='S14_3_tunneling_H_vs_D.mp4', fps=30):
    try:
        writer = FFMpegWriter(fps=fps, bitrate=2500)
        anim.save(filename_mp4, writer=writer, dpi=160)
        print(f"[OK] Saved MP4 -> {filename_mp4}")
    except Exception as e:
        print("[WARN] ffmpeg unavailable; falling back to GIF:", e)
        gif_name = filename_mp4.replace('.mp4', '.gif')
        writer = PillowWriter(fps=fps)
        anim.save(gif_name, writer=writer, dpi=160)
        print(f"[OK] Saved GIF -> {gif_name}")

# Save and show
save_animation(anim)
plt.tight_layout()
plt.show()
```

```python
# S14.3 â€” 2D Tunneling Movie (CPU-only, split-operator TDSE, H vs D side-by-side)
# Saves: S14_3_tunneling_H_vs_D.mp4  (or .gif fallback) and shows the last frame.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter

# -----------------------------
# Spatial grid & constants
# -----------------------------
Nx, Ny = 160, 120          # grid (adjust up/down if you need resolution vs speed)
Lx, Ly = 240.0, 120.0      # domain size
dx, dy = Lx / Nx, Ly / Ny
x = np.linspace(-Lx/2, Lx/2 - dx, Nx)
y = np.linspace(-Ly/2, Ly/2 - dy, Ny)
X, Y = np.meshgrid(x, y, indexing='xy')

hbar = 1.0
m_H  = 1.0                 # "light" isotope (H)
m_D  = 2.0                 # "heavy" isotope (D)

# -----------------------------
# Time stepping
# -----------------------------
dt             = 0.08
n_frames       = 260
steps_per_frame= 2

# -----------------------------
# Potential: barrier + pocket
# -----------------------------
barrier_center = 0.0
barrier_width  = 6.0
barrier_height = 18.0

pocket_x_start = 30.0
pocket_depth   = -35.0

V = np.zeros((Ny, Nx), dtype=np.float64)
# barrier slab
barrier_mask = (np.abs(X - barrier_center) < (barrier_width/2))
V[barrier_mask] += barrier_height
# reaction pocket to the right
V[X > pocket_x_start] += pocket_depth

# Soft edge walls (very weak) to discourage wrap-around
edge_soft = 24.0
V += 0.02 * (np.tanh((np.abs(X) - (Lx/2 - edge_soft))/2) + 1.0)
V += 0.02 * (np.tanh((np.abs(Y) - (Ly/2 - edge_soft))/2) + 1.0)

# -----------------------------
# Initial wavepackets
# -----------------------------
x0, y0  = -70.0, 0.0
sigx    = 7.0
sigy    = 8.0

# Make H "hotter" to show transmission; D "cooler" to show reflection
kx0_H, ky0_H = 1.55, 0.0
kx0_D, ky0_D = 1.15, 0.0

gauss = np.exp(-((X - x0)**2/(2*sigx**2) + (Y - y0)**2/(2*sigy**2)))

psi_H = gauss * np.exp(1j * (kx0_H * X + ky0_H * Y))
psi_D = gauss * np.exp(1j * (kx0_D * X + ky0_D * Y))

# normalize
def normalize(psi):
    norm = np.sqrt(np.sum(np.abs(psi)**2) * dx * dy)
    return psi / (norm if norm != 0 else 1.0)

psi_H = normalize(psi_H)
psi_D = normalize(psi_D)

# -----------------------------
# Absorbing boundaries (cosine mask)
# -----------------------------
absorb_margin = 24.0
mask_x = np.ones(Nx)
mask_y = np.ones(Ny)

# cosine turn-on near left/right
for i, xx in enumerate(x):
    if xx < -Lx/2 + absorb_margin:
        s = (xx + Lx/2) / absorb_margin
        mask_x[i] = np.sin(0.5*np.pi*max(0.0, s))**2
    elif xx >  Lx/2 - absorb_margin:
        s = (Lx/2 - xx) / absorb_margin
        mask_x[i] = np.sin(0.5*np.pi*max(0.0, s))**2

for j, yy in enumerate(y):
    if yy < -Ly/2 + absorb_margin:
        s = (yy + Ly/2) / absorb_margin
        mask_y[j] = np.sin(0.5*np.pi*max(0.0, s))**2
    elif yy >  Ly/2 - absorb_margin:
        s = (Ly/2 - yy) / absorb_margin
        mask_y[j] = np.sin(0.5*np.pi*max(0.0, s))**2

absorb_mask = np.outer(mask_y, mask_x)

# -----------------------------
# Split-operator (periodic core + absorber)
# -----------------------------
kx = 2*np.pi * np.fft.fftfreq(Nx, d=dx)
ky = 2*np.pi * np.fft.fftfreq(Ny, d=dy)
KX, KY = np.meshgrid(kx, ky, indexing='xy')

T_k_H = (hbar**2)*(KX**2 + KY**2) / (2*m_H)
T_k_D = (hbar**2)*(KX**2 + KY**2) / (2*m_D)

expV_half = np.exp(-1j * V * (dt/(2*hbar)))
expT_H    = np.exp(-1j * T_k_H * (dt/hbar))
expT_D    = np.exp(-1j * T_k_D * (dt/hbar))

def step_split_operator(psi, expT):
    psi *= expV_half
    psi = np.fft.ifft2(np.fft.fft2(psi) * expT)
    psi *= expV_half
    # apply absorber
    psi *= absorb_mask
    return psi

# -----------------------------
# Animation & plotting
# -----------------------------
fig, (axH, axD) = plt.subplots(1, 2, figsize=(11.5, 4.8), sharey=True)

imH = axH.imshow(np.abs(psi_H)**2, extent=[x.min(), x.max(), y.min(), y.max()],
                 origin='lower', interpolation='nearest', aspect='auto')
imD = axD.imshow(np.abs(psi_D)**2, extent=[x.min(), x.max(), y.min(), y.max()],
                 origin='lower', interpolation='nearest', aspect='auto')

axH.set_title('H (m=1): |Ïˆ|Â²'); axD.set_title('D (m=2): |Ïˆ|Â²')
for ax in (axH, axD):
    ax.set_xlabel('x'); ax.set_xlim(-120, 120)
axH.set_ylabel('y')

# contours (ensure increasing levels; pocket_depth < 0 < barrier_height)
levels = [pocket_depth, pocket_depth/2.0, 0.0, barrier_height/2.0, barrier_height]
for ax in (axH, axD):
    cs = ax.contour(X, Y, V, levels=levels, colors='white', linewidths=0.6, alpha=0.6)
    ax.clabel(cs, inline=True, fontsize=7, fmt='%g')

cbarH = plt.colorbar(imH, ax=axH); cbarH.set_label('Prob. density')
cbarD = plt.colorbar(imD, ax=axD); cbarD.set_label('Prob. density')

textH = axH.text(0.02, 0.97, '', transform=axH.transAxes, color='white',
                 fontsize=9, va='top', ha='left',
                 bbox=dict(facecolor='black', alpha=0.4, boxstyle='round,pad=0.2'))
textD = axD.text(0.02, 0.97, '', transform=axD.transAxes, color='white',
                 fontsize=9, va='top', ha='left',
                 bbox=dict(facecolor='black', alpha=0.4, boxstyle='round,pad=0.2'))

frame_log_marks = set(int(i*(n_frames)/20) for i in range(1, 21))  # ~5% logs

def update(frame):
    global psi_H, psi_D
    for _ in range(steps_per_frame):
        psi_H = step_split_operator(psi_H, expT_H)
        psi_D = step_split_operator(psi_D, expT_D)

    densH = np.abs(psi_H)**2
    densD = np.abs(psi_D)**2

    imH.set_data(densH); imD.set_data(densD)

    # auto-contrast separately for each panel
    imH.set_clim(0, np.percentile(densH, 99.5))
    imD.set_clim(0, np.percentile(densD, 99.5))

    t_now = dt*steps_per_frame*(frame+1)
    textH.set_text(f't = {t_now:.2f}')
    textD.set_text(f't = {t_now:.2f}')

    if frame in frame_log_marks:
        pct = int(100*(frame+1)/n_frames)
        print(f"[S14.3] {pct}%")

    return (imH, imD, textH, textD)

anim = FuncAnimation(fig, update, frames=n_frames, interval=30, blit=False)

# Save movie (MP4 preferred, GIF fallback)
def save_animation(anim, filename_mp4='S14_3_tunneling_H_vs_D.mp4', fps=30):
    try:
        writer = FFMpegWriter(fps=fps, bitrate=2500)
        anim.save(filename_mp4, writer=writer, dpi=160)
        print(f"[OK] Saved MP4 -> {filename_mp4}")
    except Exception as e:
        print("[WARN] ffmpeg unavailable; falling back to GIF:", e)
        gif_name = filename_mp4.replace('.mp4', '.gif')
        writer = PillowWriter(fps=fps)
        anim.save(gif_name, writer=writer, dpi=160)
        print(f"[OK] Saved GIF -> {gif_name}")

save_animation(anim)

plt.tight_layout()
plt.show()
```
