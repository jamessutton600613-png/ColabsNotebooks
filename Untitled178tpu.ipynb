```python
import pickle
import os
import jax
import jax.numpy as jnp
import numpy as np # Import numpy for initial state creation
import re
import jax.lax # Explicitly import jax.lax

# Define checkpoint interval
CHECKPOINT_INTERVAL = 1000 # Save checkpoint every 1000 steps

# Define the directory where checkpoints are saved - Explicitly define it here
CHECKPOINT_DIR = "/content/drive/MyDrive/Colab Notebooks/qtpu/replica_1/Expt_2025-7-6_08-50" # Ensure this matches the directory used in the simulation

# Create the directory if it doesn't exist
os.makedirs(CHECKPOINT_DIR, exist_ok=True)
print(f"Ensured checkpoint directory exists: {CHECKPOINT_DIR}")

# List all files in the checkpoint directory
all_files = os.listdir(CHECKPOINT_DIR)

# Filter for checkpoint files and sort them by step number
checkpoint_files_to_load = []
# Regex to match checkpoint files (initial_state.pkl, checkpoint_step_X.pkl, final_state_step_X.pkl)
checkpoint_pattern = re.compile(r'^(initial_state\.pkl|checkpoint_step_(\d+)\.pkl|final_state_step_(\d+)\.pkl)$')

for filename in all_files:
    match = checkpoint_pattern.match(filename)
    if match:
        file_path = os.path.join(CHECKPOINT_DIR, filename)
        # Extract step number for sorting
        if 'initial_state' in filename:
            step = 0
        elif 'final_state' in filename:
             # Assuming final state filename is final_state_step_NUM_STEPS.pkl
             # We need NUM_STEPS. Let's try to extract it from the filename if available
             try:
                 step = int(match.group(3)) # Capture group 3 for final_state_step_(\d+).pkl
             except (ValueError, TypeError):
                  # Fallback to a large number for sorting if extraction fails
                  step = 999999999
        else:
            step = int(match.group(2)) # Capture group 2 for checkpoint_step_(\d+).pkl
        checkpoint_files_to_load.append((step, file_path))

# Sort files by step number
checkpoint_files_to_load.sort(key=lambda x: x[0])

# Load the last checkpoint if any exist
population_state = None # Initialize population_state to None
start_step = 0

if checkpoint_files_to_load:
    last_step, last_file_path = checkpoint_files_to_load[-1]
    try:
        with open(last_file_path, 'rb') as f:
            population_state = pickle.load(f)
            start_step = last_step + 1
            print(f"Successfully loaded last checkpoint: {last_file_path} (Step: {last_step})")
            print(f"Resuming simulation from step {start_step} with loaded state.")
    except FileNotFoundError:
        print(f"Error: Last checkpoint file not found at {last_file_path}. Starting new simulation with re-initialized state.")
    except Exception as e:
        print(f"Error loading last checkpoint {last_file_path}: {e}. Starting new simulation with re-initialized state.")


# If population_state is still None (meaning no checkpoints were loaded or loading failed),
# re-initialize it.
if population_state is None:
    print(f"No valid checkpoints found or loaded. Starting new simulation from step {start_step} with re-initialized state.")
    # Define simulation parameters (assuming these are consistent)
    NUM_AGENTS = 50000
    SEQUENCE_LENGTH = 100

    # Define integer representations for bases (assuming these are consistent)
    BASE_A = 0
    BASE_U = 1
    BASE_G = 2
    BASE_C = 3

    # Define protoribosome states (assuming these are consistent)
    PROTO_STATE_ACTIVE = 0
    PROTO_STATE_ARRESTED = 1
    PROTO_STATE_INACTIVE = 2

    # Define damaged base representations
    DAMAGED_G = 8 # Representing 8-oxoguanine
    DAMAGED_PYRIMIDINE_DIMER = 99 # Representing pyrimidine dimers (simplified as a single type)

    # Define constants for environmental cycles (re-defined here for initial state)
    STEPS_PER_DAY = 24 # Assuming 24 steps represent one day
    DAYS_PER_YEAR = 365
    STEPS_PER_YEAR = STEPS_PER_DAY * DAYS_PER_YEAR # Ensure consistency

    # Environmental parameters (can be adjusted)
    MAX_UV_DAILY = 1.0 # Maximum UV intensity during the day
    MIN_UV_DAILY = 0.1 # Minimum UV intensity (night)
    MAX_UV_SEASONAL_FACTOR = 1.5 # UV is 1.5 times higher in summer peak
    MIN_UV_SEASONAL_FACTOR = 0.5 # UV is 0.5 times lower in winter peak
    SUMMER_SOLSTICE_DAY = DAYS_PER_YEAR // 2 # Mid-year is summer peak
    DAY_CYCLE_OFFSET = STEPS_PER_DAY // 2 # Peak UV at midday

    BASE_TEMP = 20.0 # Base temperature
    TEMP_DAILY_AMPLITUDE = 5.0 # Daily temperature swing
    TEMP_SEASONAL_AMPLITUDE = 10.0 # Seasonal temperature swing
    TEMP_DAILY_OFFSET = STEPS_PER_DAY // 3 * 2 # Peak temp in afternoon
    TEMP_SEASONAL_OFFSET = SUMMER_SOLSTICE_DAY * STEPS_PER_DAY # Corrected typo here


    # Re-initialize JAX random key for initial state if needed (using a fixed seed for reproducibility of initial state)
    initial_key = jax.random.PRNGKey(42)
    initial_key, strategy_key, rna_key, proto_key = jax.random.split(initial_key, 4)

    # Re-initialize population state structure
    initial_strategies = jax.random.randint(strategy_key, (NUM_AGENTS,), 0, 2)

    # Re-initialize RNA sequences with random canonical bases (0-3 representing A, U, G, C)
    initial_rna_sequences = np.random.randint(0, 4, size=(NUM_AGENTS, SEQUENCE_LENGTH), dtype=np.int8)

    # Re-initialize protoribosome states (e.g., all start as ACTIVE)
    initial_protoribosome_states = np.full(NUM_AGENTS, PROTO_STATE_ACTIVE, dtype=np.int8)

    # Initial state should not have damage or stop codons
    population_state = {
        'strategy': np.array(initial_strategies),
        'rna_damage': np.zeros((NUM_AGENTS, SEQUENCE_LENGTH), dtype=np.float32), # Keep general damage level for simplicity alongside damaged base type
        'has_stop_codon': jnp.zeros(NUM_AGENTS, dtype=bool),
        # 'status': np.ones(NUM_AGENTS, dtype=np.int8), # Remove status, agents are implicitly "present" unless removed
        'rna_sequence': initial_rna_sequences, # RNA sequence now stores base type OR damaged base type
        'protoribosome_state': initial_protoribosome_states,
        'protoribosome_arrest_timer': jnp.zeros(NUM_AGENTS, dtype=jnp.int32), # Initialize timer
        'autocatalysis_failure_timer': jnp.zeros(NUM_AGENTS, dtype=jnp.int32) # Timer for consecutive autocatalysis failures
    }
    start_step = 0 # Ensure start_step is 0 if re-initializing

    # Save the initial state if starting from step 0
    initial_state_filename = os.path.join(CHECKPOINT_DIR, 'initial_state.pkl')
    # Check if the initial state file already exists to avoid overwriting on accidental re-run
    if not os.path.exists(initial_state_filename):
         try:
              with open(initial_state_filename, 'wb') as f:
                  pickle.dump(population_state, f)
              print(f"Saved initial state to {initial_state_filename}")
         except Exception as e:
              print(f"Error saving initial state: {e}")
    else:
         print(f"Initial state file already exists: {initial_state_filename}. Not overwriting.")


# Define constants for environmental cycles (Ensure these are consistent with re-initialization)
STEPS_PER_DAY = 24 # Assuming 24 steps represent one day
DAYS_PER_YEAR = 365
STEPS_PER_YEAR = STEPS_PER_DAY * DAYS_PER_YEAR

# Environmental parameters (can be adjusted)
MAX_UV_DAILY = 1.0 # Maximum UV intensity during the day
MIN_UV_DAILY = 0.1 # Minimum UV intensity (night)
MAX_UV_SEASONAL_FACTOR = 1.5 # UV is 1.5 times higher in summer peak
MIN_UV_SEASONAL_FACTOR = 0.5 # UV is 0.5 times lower in winter peak
SUMMER_SOLSTICE_DAY = DAYS_PER_YEAR // 2 # Mid-year is summer peak
DAY_CYCLE_OFFSET = STEPS_PER_DAY // 2 # Peak UV at midday

BASE_TEMP = 20.0 # Base temperature
TEMP_DAILY_AMPLITUDE = 5.0 # Daily temperature swing
TEMP_SEASONAL_AMPLITUDE = 10.0 # Seasonal temperature swing
TEMP_DAILY_OFFSET = STEPS_PER_DAY // 3 * 2 # Peak temp in afternoon
TEMP_SEASONAL_OFFSET = SUMMER_SOLSTICE_DAY * STEPS_PER_DAY # Corrected typo here

# --- Simulation Parameters for Step Logic ---
# Damage rate parameters for converting canonical bases to damaged bases based on UV
UV_DAMAGE_RATE_G_TO_8OXOG = 0.001 # Probability per G per UV unit per step to become 8-oxoG
UV_DAMAGE_RATE_PYRIMIDINE_DIMER = 0.002 # Probability per adjacent pyrimidine pair per UV unit per step to become a dimer (simplified)

# Repair rate parameters for converting damaged bases back to canonical bases
REPAIR_RATE_8OXOG = 0.01 # Probability per 8-oxoG per step to be repaired
REPAIR_RATE_PYRIMIDINE_DIMER = 0.008 # Probability per dimer per step to be repaired

# Error-prone repair rate parameters
ERROR_PRONE_RATE_8OXOG = 0.1 # Probability that 8-oxoG repair is error-prone leading to mutation
ERROR_PRONE_RATE_PYRIMIDINE_DIMER = 0.15 # Probability that dimer repair is error-prone leading to mutation


# Strategy-dependent parameters
# Cautious strategy (0) has higher repair rate and lower damage threshold for arrest
CAUTIOUS_REPAIR_MULTIPLIER = 1.5 # Cautious agents repair 1.5x faster
RECKLESS_REPAIR_MULTIPLIER = 1.0 # Reckless agents repair at base rate
CAUTIOUS_ARREST_THRESHOLD_MULTIPLIER = 0.8 # Cautious agents arrest at 80% of base threshold
RECKLESS_ARREST_THRESHOLD_MULTIPLIER = 1.0 # Reckless agents arrest at base threshold

# Stop codon definitions (UAA, UAG, UGA) as integer sequences
STOP_CODON_UAA = jnp.array([BASE_U, BASE_A, BASE_A], dtype=jnp.int8) # 1, 0, 0
STOP_CODON_UAG = jnp.array([BASE_U, BASE_A, BASE_G], dtype=jnp.int8) # 1, 0, 2
STOP_CODON_UGA = jnp.array([BASE_U, BASE_G, BASE_A], dtype=jnp.int8) # 1, 2, 0

# Protoribosome state transition parameters
DAMAGE_BASE_COUNT_ARREST_THRESHOLD = 5 # Example: arrest if more than 5 damaged bases
ARREST_DURATION_STEPS = 10 # Number of steps an arrested protoribosome stays arrested before potentially becoming inactive (simplified)
REINITIATION_RATE = 0.01 # Probability per step for inactive protoribosomes to become active if conditions allow

# Autocatalysis (Reproduction/Failure) Parameters
# Probability of successful autocatalysis per step for agents with ACTIVE protoribosomes
AUTOCATALYSIS_SUCCESS_RATE_ACTIVE = 0.05
# Probability of failure per step for agents with INACTIVE protoribosomes
AUTOCATALYSIS_FAILURE_RATE_INACTIVE = 0.1
# Probability of failure per step for agents with ARRESTED protoribosomes
AUTOCATALYSIS_FAILURE_RATE_ARRESTED = 0.05
# Additional failure probability per step for agents with very high damage
AUTOCATALYSIS_FAILURE_RATE_HIGH_DAMAGE = 0.02
# Damage threshold for triggering additional failure probability (based on average damage per base)
HIGH_DAMAGE_FAILURE_THRESHOLD = 0.5 # Example threshold based on average damage per base

# Foraging/Removal Parameters
FORAGING_RATE_FAILED = 0.001 # Probability per step for a failed agent to be removed
FAILURE_DURATION_FORAGING_THRESHOLD = 50 # Number of consecutive failure steps before foraging risk increases (simplified)


# Main simulation loop
NUM_STEPS = 10000
print(f"Running simulation from step {start_step} to {NUM_STEPS}")

# Ensure population_state is not None before starting the loop
if population_state is not None:
    # Initialize random key for stochastic processes within the loop
    # Use a key derived from the step number for reproducibility across steps
    key = jax.random.PRNGKey(start_step)

    # Ensure timers are in population_state if resuming from an older checkpoint structure
    if 'protoribosome_arrest_timer' not in population_state:
         population_state['protoribosome_arrest_timer'] = jnp.zeros(NUM_AGENTS, dtype=jnp.int32)
    if 'autocatalysis_failure_timer' not in population_state:
         population_state['autocatalysis_failure_timer'] = jnp.zeros(NUM_AGENTS, dtype=jnp.int32)


    for step in range(start_step, NUM_STEPS + 1):
        # Initialize num_error_prone_mutations for this step
        num_error_prone_mutations = 0

        # Update the random key for the current step
        key, step_key = jax.random.split(key)

        # Calculate environmental variables for the current step
        current_uv_radiation = calculate_uv(step)
        current_temperature = calculate_temperature(step)

        # --- 1. Environmental Factors & RNA Damage Application (Specific Damaged Bases) ---
        # Apply UV damage to RNA sequences for all agents
        current_rna_sequence = population_state['rna_sequence']

        # Identify canonical G bases
        canonical_g_mask = (current_rna_sequence == BASE_G)

        # Identify pyrimidine bases (U or C)
        pyrimidine_mask = ((current_rna_sequence == BASE_U) | (current_rna_sequence == BASE_C))

        # Identify adjacent pyrimidine pairs (simplified: check if a pyrimidine is next to another pyrimidine)
        pyrimidine_mask_shifted = jnp.roll(pyrimidine_mask, -1, axis=1)
        pyrimidine_mask_shifted = jnp.where(jnp.arange(SEQUENCE_LENGTH) == SEQUENCE_LENGTH - 1, False, pyrimidine_mask_shifted)
        potential_dimer_site_mask = pyrimidine_mask & pyrimidine_mask_shifted

        # Determine which canonical Gs become 8-oxoG due to UV
        key, uv_g_key = jax.random.split(step_key)
        uv_g_roll = jax.random.uniform(uv_g_key, shape=canonical_g_mask.shape)
        g_to_8oxog_mask = canonical_g_mask & (uv_g_roll < current_uv_radiation * UV_DAMAGE_RATE_G_TO_8OXOG)

        # Determine which potential pyrimidine dimer sites become dimers due to UV
        key, uv_pyr_key = jax.random.split(key)
        uv_pyr_roll = jax.random.uniform(uv_pyr_key, shape=potential_dimer_site_mask.shape)
        to_pyrimidine_dimer_mask = potential_dimer_site_mask & (uv_pyr_roll < current_uv_radiation * UV_DAMAGE_RATE_PYRIMIDINE_DIMER)

        # Update RNA sequence: canonical bases become damaged bases
        new_rna_sequence = jnp.copy(current_rna_sequence)
        new_rna_sequence = jnp.where(g_to_8oxog_mask, jnp.array(DAMAGED_G, dtype=jnp.int8), new_rna_sequence)
        new_rna_sequence = jnp.where(to_pyrimidine_dimer_mask, jnp.array(DAMAGED_PYRIMIDINE_DIMER, dtype=jnp.int8), new_rna_sequence)

        population_state['rna_sequence'] = new_rna_sequence

        # --- 2. RNA Repair (Targeting Damaged Bases) ---
        # Identify damaged bases
        damaged_g_mask = (population_state['rna_sequence'] == DAMAGED_G)
        damaged_dimer_mask = (population_state['rna_sequence'] == DAMAGED_PYRIMIDINE_DIMER)

        # Determine repair probability based on strategy
        repair_rate_multiplier = jnp.where(
            population_state['strategy'] == 0, # Cautious strategy
            jnp.array(CAUTIOUS_REPAIR_MULTIPLIER),
            jnp.array(RECKLESS_REPAIR_MULTIPLIER) # Reckless strategy
        )[:, None] # Expand dims for broadcasting

        # Determine which damaged Gs and dimers are repaired
        key, repair_roll_key = jax.random.split(key) # Generate repair roll for this step
        repair_roll = jax.random.uniform(repair_roll_key, shape=damaged_g_mask.shape) # Use shape of damage mask

        g_repair_mask = damaged_g_mask & (repair_roll < REPAIR_RATE_8OXOG * repair_rate_multiplier) # Apply strategy multiplier
        dimer_repair_mask = damaged_dimer_mask & (repair_roll < REPAIR_RATE_PYRIMIDINE_DIMER * repair_rate_multiplier) # Apply strategy multiplier


        # --- 3. Implement Mutations (Error-Prone Repair of Damaged Bases) ---
        # Mutations occur when repair is error-prone.
        # Determine which repaired damaged Gs result in error-prone repair
        key, error_prone_g_key = jax.random.split(key)
        error_prone_g_roll = jax.random.uniform(error_prone_g_key, shape=g_repair_mask.shape)
        error_prone_g_repair_mask = g_repair_mask & (error_prone_g_roll < ERROR_PRONE_RATE_8OXOG)

        # Determine which repaired dimers result in error-prone repair
        key, error_prone_dimer_key = jax.random.split(key)
        error_prone_dimer_roll = jax.random.uniform(error_prone_dimer_key, shape=dimer_repair_mask.shape)
        error_prone_dimer_repair_mask = dimer_repair_mask & (error_prone_dimer_roll < ERROR_PRONE_RATE_PYRIMIDINE_DIMER)

        # Combine masks for all sites undergoing error-prone repair
        error_prone_repair_mask = error_prone_g_repair_mask | error_prone_dimer_repair_mask

        # At sites undergoing error-prone repair, the base changes to a random canonical base (0-3).
        num_error_prone_mutations = jnp.sum(error_prone_repair_mask) # Calculate before the if block

        if num_error_prone_mutations > 0:
             key, mutation_type_key = jax.random.split(key)
             # Generate random new canonical bases (0-3) for the mutation sites
             random_new_bases = jax.random.randint(mutation_type_key, shape=(num_error_prone_mutations,), minval=0, maxval=4)

             # Get the indices where mutations occur.
             mutation_indices = jnp.argwhere(error_prone_repair_mask, size=num_error_prone_mutations, fill_value=-1) # Get (agent_idx, base_idx)

             # Update the rna_sequence at the error-prone repair sites with random canonical bases
             population_state['rna_sequence'] = jax.lax.scatter_nd_update(
                 population_state['rna_sequence'],
                 mutation_indices,
                 random_new_bases,
                 scatter_dims=(0, 1) # Scatter along agent and sequence dimensions
             )

        # For damaged bases that were repaired accurately (not error-prone), they revert to their canonical form.
        # Accurate repair of 8-oxoG (8) goes back to G (2).
        # Accurate repair of dimer (99) goes back to a random pyrimidine (U=1 or C=3) for simplicity.

        accurate_g_repair_mask = g_repair_mask & ~error_prone_g_repair_mask
        accurate_dimer_repair_mask = dimer_repair_mask & ~error_prone_dimer_repair_mask

        # Apply accurate G repair: 8-oxoG (8) becomes G (2)
        population_state['rna_sequence'] = jnp.where(
            accurate_g_repair_mask,
            jnp.array(BASE_G, dtype=jnp.int8),
            population_state['rna_sequence']
        )

        # Apply accurate dimer repair: Dimer (99) becomes a random pyrimidine (U=1 or C=3)
        num_accurate_dimer_repairs = jnp.sum(accurate_dimer_repair_mask)
        if num_accurate_dimer_repairs > 0:
             key, accurate_dimer_base_key = jax.random.split(key)
             # Generate random pyrimidine bases (1 or 3)
             random_pyr_bases = jax.random.randint(accurate_dimer_base_key, shape=(num_accurate_dimer_repairs,), minval=0, maxval=2) # 0 or 1
             # Map 0 to U (1) and 1 to C (3)
             random_canonical_pyr = jnp.where(random_pyr_bases == 0, jnp.array(BASE_U, dtype=jnp.int8), jnp.array(BASE_C, dtype=jnp.int8))

             # Get the indices where accurate dimer repairs occur.
             accurate_dimer_indices = jnp.argwhere(accurate_dimer_repair_mask, size=num_accurate_dimer_repairs, fill_value=-1) # Get (agent_idx, base_idx)

             # Update the rna_sequence at the accurate dimer repair sites
             population_state['rna_sequence'] = jax.lax.scatter_nd_update(
                 population_state['rna_sequence'],
                 accurate_dimer_indices,
                 random_canonical_pyr,
                 scatter_dims=(0, 1) # Scatter along agent and sequence dimensions
             )


        # --- 4. Determine Stop Codons ---
        # Check for the presence of stop codons in the RNA sequence for each agent.
        # This needs to be done after mutations, as mutations can create or remove stop codons.
        # Damaged bases (8 and 99) are not canonical bases and would likely block translation if encountered by the protoribosome.
        # For stop codon check, we should only look for canonical bases.

        # Reset has_stop_codon for all agents before checking
        population_state['has_stop_codon'] = jnp.array(False, shape=NUM_AGENTS)

        # Check for each stop codon in each agent's RNA sequence
        # This part is not easily jittable due to variable inner loop length (sequence scanning)
        # For clarity and direct implementation of the concept, let's keep this for loop over agents for now.

        has_stop_codon_mask = jnp.zeros(NUM_AGENTS, dtype=bool)

        if SEQUENCE_LENGTH >= 3:
            for i in range(NUM_AGENTS):
                 rna = population_state['rna_sequence'][i]
                 is_stop = False
                 # Iterate through possible codon start positions
                 for j in range(SEQUENCE_LENGTH - 2):
                     codon = rna[j : j + 3]
                     # Check if all bases in the codon are canonical (0-3)
                     if jnp.all((codon >= 0) & (codon <= 3)):
                          # Check for stop codons
                         if jnp.array_equal(codon, STOP_CODON_UAA) or \
                            jnp.array_equal(codon, STOP_CODON_UAG) or \
                            jnp.array_equal(codon, STOP_CODON_UGA):
                             is_stop = True
                             break # Found a stop codon, no need to check further for this agent
                     # If any base in the codon is a damaged base (8 or 99), it's not a valid stop codon here
                     # and would likely cause protoribosome arrest (handled in the next step).

                 has_stop_codon_mask = has_stop_codon_mask.at[i].set(is_stop)

        population_state['has_stop_codon'] = has_stop_codon_mask


        # --- 5. Implement Protoribosome State Updates ---
        # Update protoribosome states based on current state, damage, and stop codons.
        current_proto_state = population_state['protoribosome_state']
        new_proto_state = jnp.copy(current_proto_state) # Start with current states

        # Calculate number of damaged bases per agent
        num_damaged_bases_per_agent = jnp.sum(
            (population_state['rna_sequence'] == DAMAGED_G) | (population_state['rna_sequence'] == DAMAGED_PYRIMIDINE_DIMER),
            axis=1
        )

        # Determine damage arrest threshold based on strategy
        damage_arrest_threshold_per_agent = jnp.where(
            population_state['strategy'] == 0, # Cautious strategy
            jnp.array(DAMAGE_BASE_COUNT_ARREST_THRESHOLD * CAUTIOUS_ARREST_THRESHOLD_MULTIPLIER),
            jnp.array(DAMAGE_BASE_COUNT_ARREST_THRESHOLD * RECKLESS_ARREST_THRESHOLD_MULTIPLIER) # Reckless strategy
        )


        # Transition rules:
        # ACTIVE (0) -> ARRESTED (1): if has_stop_codon is True OR num_damaged_bases_per_agent > strategy-dependent threshold
        # ARRESTED (1) -> INACTIVE (2) : if protoribosome_arrest_timer exceeds ARREST_DURATION_STEPS
        # ARRESTED (1) -> ACTIVE (0): if damaged bases are repaired AND stop codon removed (simplified: if no stop codon AND num_damaged_bases_per_agent <= strategy-dependent threshold)
        # INACTIVE (2) -> ACTIVE (0): (Optional, re-initiation) after a period of inactivity (e.g., if timer exceeds a threshold) AND conditions for arrest are no longer met

        # Update protoribosome arrest timer for agents currently in ARRESTED state
        arrested_mask = (current_proto_state == PROTO_STATE_ARRESTED)
        protoribosome_arrest_timer = jnp.where(
            arrested_mask,
            population_state['protoribosome_arrest_timer'] + 1, # Use protoribosome_arrest_timer from population_state
            jnp.array(0, dtype=jnp.int32) # Timer resets for other states
        )

        # Transition from ACTIVE to ARRESTED
        active_to_arrested_mask = (current_proto_state == PROTO_STATE_ACTIVE) & \
                                   (population_state['has_stop_codon'] | (num_damaged_bases_per_agent > damage_arrest_threshold_per_agent))

        new_proto_state = jnp.where(
            active_to_arrested_mask,
            jnp.array(PROTO_STATE_ARRESTED, dtype=jnp.int8), # Ensure dtype matches population_state
            new_proto_state
        )


        # Transition from ARRESTED to INACTIVE
        # This transition is based on the arrest timer
        arrested_to_inactive_mask = (current_proto_state == PROTO_STATE_ARRESTED) & \
                                      (protoribosome_arrest_timer >= jnp.array(ARREST_DURATION_STEPS))

        new_proto_state = jnp.where(
            arrested_to_inactive_mask,
            jnp.array(PROTO_STATE_INACTIVE, dtype=jnp.int8), # Ensure dtype matches population_state
            new_proto_state
        )

        # Transition from ARRESTED back to ACTIVE (Recovery)
        # Conditions: No stop codon AND damage is below the strategy-dependent threshold
        arrested_to_active_mask = (current_proto_state == PROTO_STATE_ARRESTED) & \
                                  (~population_state['has_stop_codon'] & (num_damaged_bases_per_agent <= damage_arrest_threshold_per_agent))

        new_proto_state = jnp.where(
            arrested_to_active_mask,
            jnp.array(PROTO_STATE_ACTIVE, dtype=jnp.int8),
            new_proto_state
        )
        # Reset timer for agents transitioning from ARRESTED to ACTIVE
        protoribosome_arrest_timer = jnp.where(
            arrested_to_active_mask,
            jnp.array(0, dtype=jnp.int32),
            protoribosome_arrest_timer
        )


        # Transition from INACTIVE back to ACTIVE (Re-initiation)
        # Let's add a simple re-initiation probability for inactive protoribosomes
        # if conditions for arrest are not met.
        inactive_to_active_mask_conditions = (current_proto_state == PROTO_STATE_INACTIVE) & \
                                             (~population_state['has_stop_codon'] & (num_damaged_bases_per_agent <= damage_arrest_threshold_per_agent))

        # Add a re-initiation probability
        key, reinit_roll_key = jax.random.split(key)
        reinit_roll = jax.random.uniform(reinit_roll_key, shape=inactive_to_active_mask_conditions.shape)
        inactive_to_active_mask = inactive_to_active_mask_conditions & (reinit_roll < REINITIATION_RATE)


        new_proto_state = jnp.where(
            inactive_to_active_mask,
            jnp.array(PROTO_STATE_ACTIVE, dtype=jnp.int8),
            new_proto_state
        )
        # Reset timer for agents transitioning from INACTIVE to ACTIVE (timer is already 0)


        # Update the protoribosome_state in the population_state
        population_state['protoribosome_state'] = new_proto_state
        # Update the arrest timer in the population_state
        population_state['protoribosome_arrest_timer'] = protoribosome_arrest_timer


        # --- 7. Implement Autocatalysis (Success/Failure and Population Dynamics) ---
        # Determine which agents are capable of successful autocatalysis (reproduction)
        # Only agents with ACTIVE protoribosomes can potentially reproduce
        can_reproduce_mask = (population_state['protoribosome_state'] == PROTO_STATE_ACTIVE)
        key, success_roll_key = jax.random.split(key)
        success_roll = jax.random.uniform(success_roll_key, shape=can_reproduce_mask.shape)
        successfully_reproducing_agents_mask = can_reproduce_mask & (success_roll < AUTOCATALYSIS_SUCCESS_RATE_ACTIVE)

        num_successfully_reproducing_agents = jnp.sum(successfully_reproducing_agents_mask)

        # Update autocatalysis failure timer: increment for agents that did NOT reproduce, reset for those that did.
        autocatalysis_failure_timer = jnp.where(
            successfully_reproducing_agents_mask,
            jnp.array(0, dtype=jnp.int32), # Reset timer on successful reproduction
            population_state['autocatalysis_failure_timer'] + 1 # Increment timer on failure to reproduce
        )
        population_state['autocatalysis_failure_timer'] = autocatalysis_failure_timer


        # --- 8. Implement Foraging/Removal ---
        # Agents can be removed from the population (foraged) if they fail autocatalysis
        # consistently for a certain period, or based on other criteria.
        # Use the autocatalysis_failure_timer to determine foraging risk.

        # Identify agents at risk of foraging (failed for a certain duration)
        foraging_risk_mask = (population_state['autocatalysis_failure_timer'] >= jnp.array(FAILURE_DURATION_FORAGING_THRESHOLD))

        # Determine which agents are foraged
        key, foraging_roll_key = jax.random.split(key)
        foraging_roll = jax.random.uniform(foraging_roll_key, shape=foraging_risk_mask.shape)
        foraged_agents_mask = foraging_risk_mask & (foraging_roll < FORAGING_RATE_FAILED)

        num_foraged_agents = jnp.sum(foraged_agents_mask)

        # Implement Population Dynamics: Replace foraged agents with offspring from successfully reproducing agents
        # This maintains a constant population size (NUM_AGENTS).
        num_replacements_needed = num_foraged_agents
        num_potential_offspring = num_successfully_reproducing_agents

        # If there are reproducing agents and replacements needed, select reproducing agents
        if num_potential_offspring > 0 and num_replacements_needed > 0:
             key, offspring_selection_key = jax.random.split(key)
             # Select indices of successfully reproducing agents (potential parents)
             reproducing_indices = jnp.where(successfully_reproducing_agents_mask)[0]

             # Randomly select parents to produce offspring to replace foraged agents
             # Use replacement=True if num_replacements_needed > num_potential_offspring
             replace_with_replacement = num_replacements_needed > num_potential_offspring
             parent_indices_for_offspring = jax.random.choice(
                 offspring_selection_key,
                 reproducing_indices,
                 shape=(num_replacements_needed,),
                 replace=replace_with_replacement
             )

             # Get the state of the selected parent agents
             parent_rna_sequences = population_state['rna_sequence'][parent_indices_for_offspring]
             parent_strategies = population_state['strategy'][parent_indices_for_offspring]

             # Create offspring state (e.g., copy parent's RNA and strategy, reset damage/protoribosome state, reset timers)
             offspring_rna_sequences = jnp.copy(parent_rna_sequences) # Offspring inherit RNA
             offspring_strategies = jnp.copy(parent_strategies) # Offspring inherit strategy
             offspring_rna_damage = jnp.zeros_like(offspring_rna_sequences, dtype=jnp.float32) # Offspring start with no general damage
             # Important: Stop codons need to be re-checked for offspring based on inherited sequence in the next step
             offspring_has_stop_codon = jnp.zeros(num_replacements_needed, dtype=bool)
             offspring_protoribosome_state = jnp.full(num_replacements_needed, PROTO_STATE_ACTIVE, dtype=jnp.int8) # Offspring protoribosomes are active
             offspring_protoribosome_arrest_timer = jnp.zeros(num_replacements_needed, dtype=jnp.int32) # Offspring timer is 0
             offspring_autocatalysis_failure_timer = jnp.zeros(num_replacements_needed, dtype=jnp.int32) # Offspring timer is 0


             # Select indices of foraged agents to be replaced
             foraged_indices = jnp.where(foraged_agents_mask)[0]

             # Update the state of the foraged agents with the offspring state
             # Use jax.lax.scatter_nd_update to update multiple indices at once
             population_state['strategy'] = jax.lax.scatter_nd_update(
                 population_state['strategy'],
                 foraged_indices[:, None], # Indices need to be shape (N, 1) for scatter
                 offspring_strategies,
                 scatter_dims=(0,) # Scatter along the first dimension (agents)
             )
             population_state['rna_damage'] = jax.lax.scatter_nd_update(
                 population_state['rna_damage'],
                 foraged_indices[:, None, None], # Indices need to be shape (N, 1, 1) for scatter into 2D array
                 offspring_rna_damage,
                 scatter_dims=(0,)
             )
             population_state['has_stop_codon'] = jax.lax.scatter_nd_update(
                 population_state['has_stop_codon'],
                 foraged_indices[:, None],
                 offspring_has_stop_codon,
                 scatter_dims=(0,)
             )
             population_state['rna_sequence'] = jax.lax.scatter_nd_update(
                 population_state['rna_sequence'],
                 foraged_indices[:, None, None], # Indices need to be shape (N, 1, 1) for scatter into 2D array
                 offspring_rna_sequences,
                 scatter_dims=(0,)
             )
             population_state['protoribosome_state'] = jax.lax.scatter_nd_update(
                 population_state['protoribosome_state'],
                 foraged_indices[:, None],
                 offspring_protoribosome_state,
                 scatter_dims=(0,)
             )
             population_state['protoribosome_arrest_timer'] = jax.lax.scatter_nd_update(
                 population_state['protoribosome_arrest_timer'],
                 foraged_indices[:, None],
                 offspring_protoribosome_arrest_timer,
                 scatter_dims=(0,)
             )
             population_state['autocatalysis_failure_timer'] = jax.lax.scatter_nd_update(
                 population_state['autocatalysis_failure_timer'],
                 foraged_indices[:, None],
                 offspring_autocatalysis_failure_timer,
                 scatter_dims=(0,)
             )

        # If no reproducing agents but foraging occurs, these spots are effectively removed (status=0 implicitly, or just replaced with empty/dummy state)
        # In this replacement model, if num_potential_offspring is 0, num_replacements_needed will also be 0 as no offspring can be created.
        # If num_successfully_reproducing_agents < num_foraged_agents, we replace as many as possible from reproducing agents.
        # The remaining foraged spots are effectively empty until future steps produce more offspring than failures.


        # --- Placeholder for other simulation step logic ---
        # (All core simulation step logic is now implemented)


        # Placeholder print statement
        if step % 1000 == 0 or step == start_step or step == NUM_STEPS:
            # Calculate average damaged bases per agent for printing
            num_damaged_bases_per_agent = jnp.sum(
                (population_state['rna_sequence'] == DAMAGED_G) | (population_state['rna_sequence'] == DAMAGED_PYRIMIDINE_DIMER),
                axis=1
            )
            avg_damaged_bases_for_print = jnp.mean(num_damaged_bases_per_agent) if num_damaged_bases_per_agent.shape[0] > 0 else 0.0

            num_agents_with_stop_codon = jnp.sum(population_state['has_stop_codon']) # Count agents with stop codons
            # Count protoribosome states for all agents
            active_protos_count = jnp.sum(population_state['protoribosome_state'] == PROTO_STATE_ACTIVE)
            arrested_protos_count = jnp.sum(population_state['protoribosome_state'] == PROTO_STATE_ARRESTED)
            inactive_protos_count = jnp.sum(population_state['protoribosome_state'] == PROTO_STATE_INACTIVE)

            # Count agents by strategy
            cautious_count = jnp.sum(population_state['strategy'] == 0)
            reckless_count = jnp.sum(population_state['strategy'] == 1)

            # Total agents remain constant at NUM_AGENTS in this replacement model


            print(f"Simulating step {step}/{NUM_STEPS}, UV: {current_uv_radiation:.2f}, Temp: {current_temperature:.0f}C, Avg Damaged Bases: {avg_damaged_bases_for_print:.2f}, Error-Prone Mutations: {num_error_prone_mutations}, Agents with Stop Codon: {num_agents_with_stop_codon}")
            print(f"  Protoribosome States: Active={active_protos_count}, Arrested={arrested_protos_count}, Inactive={inactive_protos_count}")
            print(f"  Agents by Strategy: Cautious={cautious_count}, Reckless={reckless_count}")
            print(f"  Autocatalysis: Successful={num_successfully_reproducing_agents}, Failed (this step): {NUM_AGENTS - num_successfully_reproducing_agents}, Foraged: {num_foraged_agents}")


        # Add checkpointing logic
        # Only save if population_state is defined
        if population_state is not None and step > 0 and (step % CHECKPOINT_INTERVAL == 0 or step == NUM_STEPS):
            if step == NUM_STEPS:
                 checkpoint_filename = os.path.join(CHECKPOINT_DIR, f'final_state_step_{step}.pkl')
            else:
                 checkpoint_filename = os.path.join(CHECKPOINT_DIR, f'checkpoint_step_{step}.pkl')

            try:
                with open(checkpoint_filename, 'wb') as f:
                    pickle.dump(population_state, f)
                print(f"Checkpoint saved at step {step} to {checkpoint_filename}")
            except Exception as e:
                print(f"Error saving checkpoint at step {step}: {e}")

else:
    print("Simulation loop skipped because population_state is not defined.")


print("Simulation loop finished.")
```

```python
import pickle
import os
import jax
import jax.numpy as jnp
import numpy as np # Import numpy for initial state creation
import re
import jax.lax # Explicitly import jax.lax

# Define simulation parameters (Always defined)
NUM_AGENTS = 50000
SEQUENCE_LENGTH = 100

# Define integer representations for bases (Always defined)
BASE_A = 0
BASE_U = 1
BASE_G = 2
BASE_C = 3

# Define protoribosome states (Always defined)
PROTO_STATE_ACTIVE = 0
PROTO_STATE_ARRESTED = 1
PROTO_STATE_INACTIVE = 2

# Define damaged base representations (Always defined)
DAMAGED_G = 8 # Representing 8-oxoguanine
DAMAGED_PYRIMIDINE_DIMER = 99 # Representing pyrimidine dimers (simplified as a single type)

# Define constants for environmental cycles (Always defined)
STEPS_PER_DAY = 24 # Assuming 24 steps represent one day
DAYS_PER_YEAR = 365
STEPS_PER_YEAR = STEPS_PER_DAY * DAYS_PER_YEAR

# Environmental parameters (can be adjusted, Always defined)
MAX_UV_DAILY = 1.0 # Maximum UV intensity during the day
MIN_UV_DAILY = 0.1 # Minimum UV intensity (night)
MAX_UV_SEASONAL_FACTOR = 1.5 # UV is 1.5 times higher in summer peak
MIN_UV_SEASONAL_FACTOR = 0.5 # UV is 0.5 times lower in winter peak
SUMMER_SOLSTICE_DAY = DAYS_PER_YEAR // 2 # Mid-year is summer peak
DAY_CYCLE_OFFSET = STEPS_PER_DAY // 2 # Peak UV at midday

BASE_TEMP = 20.0 # Base temperature
TEMP_DAILY_AMPLITUDE = 5.0 # Daily temperature swing
TEMP_SEASONAL_AMPLITUDE = 10.0 # Seasonal temperature swing
TEMP_DAILY_OFFSET = STEPS_PER_DAY // 3 * 2 # Peak temp in afternoon
TEMP_SEASONAL_OFFSET = SUMMER_SOLSTICE_DAY * STEPS_PER_DAY


# Define checkpoint interval
CHECKPOINT_INTERVAL = 1000 # Save checkpoint every 1000 steps

# Define the directory where checkpoints are saved - Explicitly define it here
CHECKPOINT_DIR = "/content/drive/MyDrive/Colab Notebooks/qtpu/replica_1/Expt_2025-7-6_08-50" # Ensure this matches the directory used in the simulation

# Create the directory if it doesn't exist
os.makedirs(CHECKPOINT_DIR, exist_ok=True)
print(f"Ensured checkpoint directory exists: {CHECKPOINT_DIR}")

# List all files in the checkpoint directory
all_files = os.listdir(CHECKPOINT_DIR)

# Filter for checkpoint files and sort them by step number
checkpoint_files_to_load = []
# Regex to match checkpoint files (initial_state.pkl, checkpoint_step_X.pkl, final_state_step_X.pkl)
checkpoint_pattern = re.compile(r'^(initial_state\.pkl|checkpoint_step_(\d+)\.pkl|final_state_step_(\d+)\.pkl)$')

for filename in all_files:
    match = checkpoint_pattern.match(filename)
    if match:
        file_path = os.path.join(CHECKPOINT_DIR, filename)
        # Extract step number for sorting
        if 'initial_state' in filename:
            step = 0
        elif 'final_state' in filename:
             # Assuming final state filename is final_state_step_NUM_STEPS.pkl
             try:
                 step = int(match.group(3)) # Capture group 3 for final_state_step_(\d+).pkl
             except (ValueError, TypeError):
                  # Fallback to a large number for sorting if extraction fails
                  step = 999999999
        else:
            step = int(match.group(2)) # Capture group 2 for checkpoint_step_(\d+).pkl
        checkpoint_files_to_load.append((step, file_path))

# Sort files by step number
checkpoint_files_to_load.sort(key=lambda x: x[0])

# Load the last checkpoint if any exist
population_state = None # Initialize population_state to None
start_step = 0

if checkpoint_files_to_load:
    last_step, last_file_path = checkpoint_files_to_load[-1]
    try:
        with open(last_file_path, 'rb') as f:
            population_state = pickle.load(f)
            start_step = last_step + 1
            print(f"Successfully loaded last checkpoint: {last_file_path} (Step: {last_step})")
            print(f"Resuming simulation from step {start_step} with loaded state.")
    except FileNotFoundError:
        print(f"Error: Last checkpoint file not found at {last_file_path}. Starting new simulation with re-initialized state.")
    except Exception as e:
        print(f"Error loading last checkpoint {last_file_path}: {e}. Starting new simulation with re-initialized state.")


# If population_state is still None (meaning no checkpoints were loaded or loading failed),
# re-initialize it.
if population_state is None:
    print(f"No valid checkpoints found or loaded. Starting new simulation from step {start_step} with re-initialized state.")
    # Note: Simulation parameters, base definitions, protoribosome states,
    # damaged base definitions, and environmental constants are now defined outside this block.

    # Re-initialize JAX random key for initial state if needed (using a fixed seed for reproducibility of initial state)
    initial_key = jax.random.PRNGKey(42)
    initial_key, strategy_key, rna_key, proto_key = jax.random.split(initial_key, 4)

    # Re-initialize population state structure
    initial_strategies = jax.random.randint(strategy_key, (NUM_AGENTS,), 0, 2)

    # Re-initialize RNA sequences with random canonical bases (0-3 representing A, U, G, C)
    initial_rna_sequences = np.random.randint(0, 4, size=(NUM_AGENTS, SEQUENCE_LENGTH), dtype=np.int8)

    # Re-initialize protoribosome states (e.g., all start as ACTIVE)
    initial_protoribosome_states = np.full(NUM_AGENTS, PROTO_STATE_ACTIVE, dtype=np.int8)

    # Initial state should not have damage or stop codons
    population_state = {
        'strategy': np.array(initial_strategies),
        'rna_damage': np.zeros((NUM_AGENTS, SEQUENCE_LENGTH), dtype=np.float32), # Keep general damage level for simplicity alongside damaged base type
        'has_stop_codon': jnp.zeros(NUM_AGENTS, dtype=bool),
        # 'status': np.ones(NUM_AGENTS, dtype=np.int8), # Remove status, agents are implicitly "present" unless removed
        'rna_sequence': initial_rna_sequences, # RNA sequence now stores base type OR damaged base type
        'protoribosome_state': initial_protoribosome_states,
        'protoribosome_arrest_timer': jnp.zeros(NUM_AGENTS, dtype=jnp.int32), # Initialize timer
        'autocatalysis_failure_timer': jnp.zeros(NUM_AGENTS, dtype=jnp.int32) # Timer for consecutive autocatalysis failures
    }
    start_step = 0 # Ensure start_step is 0 if re-initializing

    # Save the initial state if starting from step 0
    initial_state_filename = os.path.join(CHECKPOINT_DIR, 'initial_state.pkl')
    # Check if the initial state file already exists to avoid overwriting on accidental re-run
    if not os.path.exists(initial_state_filename):
         try:
              with open(initial_state_filename, 'wb') as f:
                  pickle.dump(population_state, f)
              print(f"Saved initial state to {initial_state_filename}")
         except Exception as e:
              print(f"Error saving initial state: {e}")
    else:
         print(f"Initial state file already exists: {initial_state_filename}. Not overwriting.")


# Function to calculate UV radiation based on step number
def calculate_uv(step, max_uv_daily=MAX_UV_DAILY, min_uv_daily=MIN_UV_DAILY,
                 max_uv_seasonal_factor=MAX_UV_SEASONAL_FACTOR, min_uv_seasonal_factor=MIN_UV_SEASONAL_FACTOR,
                 steps_per_day=STEPS_PER_DAY, steps_per_year=STEPS_PER_YEAR,
                 day_cycle_offset=DAY_CYCLE_OFFSET):
    """Calculates UV radiation based on daily and seasonal cycles."""
    # Daily cycle (sinusoidal, peaking around midday)
    day_of_year_step = step % steps_per_year
    step_of_day = step % steps_per_day
    daily_uv = ((max_uv_daily - min_uv_daily) / 2) * (1 + jnp.cos((2 * jnp.pi / steps_per_day) * (step_of_day - day_cycle_offset))) + min_uv_daily

    # Seasonal cycle (sinusoidal, peaking in summer)
    seasonal_factor = ((max_uv_seasonal_factor - min_uv_seasonal_factor) / 2) * (1 + jnp.cos((2 * jnp.pi / steps_per_year) * (day_of_year_step - TEMP_SEASONAL_OFFSET))) + min_uv_seasonal_factor

    # Combine daily and seasonal cycles
    current_uv = daily_uv * seasonal_factor
    return current_uv

# Function to calculate temperature based on step number
def calculate_temperature(step, base_temp=BASE_TEMP, temp_daily_amplitude=TEMP_DAILY_AMPLITUDE,
                          temp_seasonal_amplitude=TEMP_SEASONAL_AMPLITUDE, steps_per_day=STEPS_PER_DAY,
                          steps_per_year=STEPS_PER_YEAR, temp_daily_offset=TEMP_DAILY_OFFSET,
                          temp_seasonal_offset=TEMP_SEASONAL_OFFSET):
    """Calculates temperature based on daily and seasonal cycles."""
    # Daily cycle (sinusoidal, peaking in afternoon)
    step_of_day = step % steps_per_day
    daily_temp = temp_daily_amplitude * jnp.sin((2 * jnp.pi / steps_per_day) * (step_of_day - temp_daily_offset))

    # Seasonal cycle (sinusoidal, peaking in summer)
    day_of_year_step = step % steps_per_year
    seasonal_temp = temp_seasonal_amplitude * jnp.sin((2 * jnp.pi / steps_per_year) * (day_of_year_step - temp_seasonal_offset))

    # Combine base temperature, daily, and seasonal cycles
    current_temp = base_temp + daily_temp + seasonal_temp
    return current_temp


# --- Simulation Parameters for Step Logic ---
# Damage rate parameters for converting canonical bases to damaged bases based on UV
UV_DAMAGE_RATE_G_TO_8OXOG = 0.001 # Probability per G per UV unit per step to become 8-oxoG
UV_DAMAGE_RATE_PYRIMIDINE_DIMER = 0.002 # Probability per adjacent pyrimidine pair per UV unit per step to become a dimer (simplified)

# Repair rate parameters for converting damaged bases back to canonical bases
REPAIR_RATE_8OXOG = 0.01 # Probability per 8-oxoG per step to be repaired
REPAIR_RATE_PYRIMIDINE_DIMER = 0.008 # Probability per dimer per step to be repaired

# Error-prone repair rate parameters
ERROR_PRONE_RATE_8OXOG = 0.1 # Probability that 8-oxoG repair is error-prone leading to mutation
ERROR_PRONE_RATE_PYRIMIDINE_DIMER = 0.15 # Probability that dimer repair is error-prone leading to mutation


# Strategy-dependent parameters
# Cautious strategy (0) has higher repair rate and lower damage threshold for arrest
CAUTIOUS_REPAIR_MULTIPLIER = 1.5 # Cautious agents repair 1.5x faster
RECKLESS_REPAIR_MULTIPLIER = 1.0 # Reckless agents repair at base rate
CAUTIOUS_ARREST_THRESHOLD_MULTIPLIER = 0.8 # Cautious agents arrest at 80% of base threshold
RECKLESS_ARREST_THRESHOLD_MULTIPLIER = 1.0 # Reckless agents arrest at base threshold

# Stop codon definitions (UAA, UAG, UGA) as integer sequences
STOP_CODON_UAA = jnp.array([BASE_U, BASE_A, BASE_A], dtype=jnp.int8) # 1, 0, 0
STOP_CODON_UAG = jnp.array([BASE_U, BASE_A, BASE_G], dtype=jnp.int8) # 1, 0, 2
STOP_CODON_UGA = jnp.array([BASE_U, BASE_G, BASE_A], dtype=jnp.int8) # 1, 2, 0

# Protoribosome state transition parameters
DAMAGE_BASE_COUNT_ARREST_THRESHOLD = 5 # Example: arrest if more than 5 damaged bases
ARREST_DURATION_STEPS = 10 # Number of steps an arrested protoribosome stays arrested before potentially becoming inactive (simplified)
REINITIATION_RATE = 0.01 # Probability per step for inactive protoribosomes to become active if conditions allow

# Autocatalysis (Reproduction/Failure) Parameters
# Probability of successful autocatalysis per step for agents with ACTIVE protoribosomes
AUTOCATALYSIS_SUCCESS_RATE_ACTIVE = 0.05
# Probability of failure per step for agents with INACTIVE protoribosomes
AUTOCATALYSIS_FAILURE_RATE_INACTIVE = 0.1
# Probability of failure per step for agents with ARRESTED protoribosomes
AUTOCATALYSIS_FAILURE_RATE_ARRESTED = 0.05
# Additional failure probability per step for agents with very high damage
AUTOCATALYSIS_FAILURE_RATE_HIGH_DAMAGE = 0.02
# Damage threshold for triggering additional failure probability (based on average damage per base)
HIGH_DAMAGE_FAILURE_THRESHOLD = 0.5 # Example threshold based on average damage per base

# Foraging/Removal Parameters
FORAGING_RATE_FAILED = 0.001 # Probability per step for a failed agent to be removed
FAILURE_DURATION_FORAGING_THRESHOLD = 50 # Number of consecutive failure steps before foraging risk increases (simplified)


# Main simulation loop
NUM_STEPS = 10000
print(f"Running simulation from step {start_step} to {NUM_STEPS}")

# Ensure population_state is not None before starting the loop
if population_state is not None:
    # Initialize random key for stochastic processes within the loop
    # Use a key derived from the step number for reproducibility across steps
    key = jax.random.PRNGKey(start_step)

    # Ensure timers are in population_state if resuming from an older checkpoint structure
    if 'protoribosome_arrest_timer' not in population_state:
         population_state['protoribosome_arrest_timer'] = jnp.zeros(NUM_AGENTS, dtype=jnp.int32)
    if 'autocatalysis_failure_timer' not in population_state:
         population_state['autocatalysis_failure_timer'] = jnp.zeros(NUM_AGENTS, dtype=jnp.int32)


    for step in range(start_step, NUM_STEPS + 1):
        # Initialize num_error_prone_mutations for this step
        num_error_prone_mutations = 0

        # Update the random key for the current step
        key, step_key = jax.random.split(key)

        # Calculate environmental variables for the current step
        current_uv_radiation = calculate_uv(step)
        current_temperature = calculate_temperature(step)

        # --- 1. Environmental Factors & RNA Damage Application (Specific Damaged Bases) ---
        # Apply UV damage to RNA sequences for all agents
        current_rna_sequence = population_state['rna_sequence']

        # Identify canonical G bases
        canonical_g_mask = (current_rna_sequence == BASE_G)

        # Identify pyrimidine bases (U or C)
        pyrimidine_mask = ((current_rna_sequence == BASE_U) | (current_rna_sequence == BASE_C))

        # Identify adjacent pyrimidine pairs (simplified: check if a pyrimidine is next to another pyrimidine)
        pyrimidine_mask_shifted = jnp.roll(pyrimidine_mask, -1, axis=1)
        pyrimidine_mask_shifted = jnp.where(jnp.arange(SEQUENCE_LENGTH) == SEQUENCE_LENGTH - 1, False, pyrimidine_mask_shifted)
        potential_dimer_site_mask = pyrimidine_mask & pyrimidine_mask_shifted

        # Determine which canonical Gs become 8-oxoG due to UV
        key, uv_g_key = jax.random.split(step_key)
        uv_g_roll = jax.random.uniform(uv_g_key, shape=canonical_g_mask.shape)
        g_to_8oxog_mask = canonical_g_mask & (uv_g_roll < current_uv_radiation * UV_DAMAGE_RATE_G_TO_8OXOG)

        # Determine which potential pyrimidine dimer sites become dimers due to UV
        key, uv_pyr_key = jax.random.split(key)
        uv_pyr_roll = jax.random.uniform(uv_pyr_key, shape=potential_dimer_site_mask.shape)
        to_pyrimidine_dimer_mask = potential_dimer_site_mask & (uv_pyr_roll < current_uv_radiation * UV_DAMAGE_RATE_PYRIMIDINE_DIMER)

        # Update RNA sequence: canonical bases become damaged bases
        new_rna_sequence = jnp.copy(current_rna_sequence)
        new_rna_sequence = jnp.where(g_to_8oxog_mask, jnp.array(DAMAGED_G, dtype=jnp.int8), new_rna_sequence)
        new_rna_sequence = jnp.where(to_pyrimidine_dimer_mask, jnp.array(DAMAGED_PYRIMIDINE_DIMER, dtype=jnp.int8), new_rna_sequence)

        population_state['rna_sequence'] = new_rna_sequence

        # --- 2. RNA Repair (Targeting Damaged Bases) ---
        # Identify damaged bases
        damaged_g_mask = (population_state['rna_sequence'] == DAMAGED_G)
        damaged_dimer_mask = (population_state['rna_sequence'] == DAMAGED_PYRIMIDINE_DIMER)

        # Determine repair probability based on strategy
        repair_rate_multiplier = jnp.where(
            population_state['strategy'] == 0, # Cautious strategy
            jnp.array(CAUTIOUS_REPAIR_MULTIPLIER),
            jnp.array(RECKLESS_REPAIR_MULTIPLIER) # Reckless strategy
        )[:, None] # Expand dims for broadcasting

        # Determine which damaged Gs and dimers are repaired
        key, repair_roll_key = jax.random.split(key) # Generate repair roll for this step
        repair_roll = jax.random.uniform(repair_roll_key, shape=damaged_g_mask.shape) # Use shape of damage mask

        g_repair_mask = damaged_g_mask & (repair_roll < REPAIR_RATE_8OXOG * repair_rate_multiplier) # Apply strategy multiplier
        dimer_repair_mask = damaged_dimer_mask & (repair_roll < REPAIR_RATE_PYRIMIDINE_DIMER * repair_rate_multiplier) # Apply strategy multiplier


        # --- 3. Implement Mutations (Error-Prone Repair of Damaged Bases) ---
        # Mutations occur when repair is error-prone.
        # Determine which repaired damaged Gs result in error-prone repair
        key, error_prone_g_key = jax.random.split(key)
        error_prone_g_roll = jax.random.uniform(error_prone_g_key, shape=g_repair_mask.shape)
        error_prone_g_repair_mask = g_repair_mask & (error_prone_g_roll < ERROR_PRONE_RATE_8OXOG)

        # Determine which repaired dimers result in error-prone repair
        key, error_prone_dimer_key = jax.random.split(key)
        error_prone_dimer_roll = jax.random.uniform(error_prone_dimer_key, shape=dimer_repair_mask.shape)
        error_prone_dimer_repair_mask = dimer_repair_mask & (error_prone_dimer_roll < ERROR_PRONE_RATE_PYRIMIDINE_DIMER)

        # Combine masks for all sites undergoing error-prone repair
        error_prone_repair_mask = error_prone_g_repair_mask | error_prone_dimer_repair_mask

        # At sites undergoing error-prone repair, the base changes to a random canonical base (0-3).
        num_error_prone_mutations = jnp.sum(error_prone_repair_mask) # Calculate before the if block

        if num_error_prone_mutations > 0:
             key, mutation_type_key = jax.random.split(key)
             # Generate random new canonical bases (0-3) for the mutation sites
             random_new_bases = jax.random.randint(mutation_type_key, shape=(num_error_prone_mutations,), minval=0, maxval=4)

             # Get the indices where mutations occur.
             mutation_indices = jnp.argwhere(error_prone_repair_mask, size=num_error_prone_mutations, fill_value=-1) # Get (agent_idx, base_idx)

             # Update the rna_sequence at the error-prone repair sites with random canonical bases
             population_state['rna_sequence'] = jax.lax.scatter_nd_update(
                 population_state['rna_sequence'],
                 mutation_indices,
                 random_new_bases,
                 scatter_dims=(0, 1) # Scatter along agent and sequence dimensions
             )

        # For damaged bases that were repaired accurately (not error-prone), they revert to their canonical form.
        # Accurate repair of 8-oxoG (8) goes back to G (2).
        # Accurate repair of dimer (99) goes back to a random pyrimidine (U=1 or C=3) for simplicity.

        accurate_g_repair_mask = g_repair_mask & ~error_prone_g_repair_mask
        accurate_dimer_repair_mask = dimer_repair_mask & ~error_prone_dimer_repair_mask

        # Apply accurate G repair: 8-oxoG (8) becomes G (2)
        population_state['rna_sequence'] = jnp.where(
            accurate_g_repair_mask,
            jnp.array(BASE_G, dtype=jnp.int8),
            population_state['rna_sequence']
        )

        # Apply accurate dimer repair: Dimer (99) becomes a random pyrimidine (U=1 or C=3)
        num_accurate_dimer_repairs = jnp.sum(accurate_dimer_repair_mask)
        if num_accurate_dimer_repairs > 0:
             key, accurate_dimer_base_key = jax.random.split(key)
             # Generate random pyrimidine bases (1 or 3)
             random_pyr_bases = jax.random.randint(accurate_dimer_base_key, shape=(num_accurate_dimer_repairs,), minval=0, maxval=2) # 0 or 1
             # Map 0 to U (1) and 1 to C (3)
             random_canonical_pyr = jnp.where(random_pyr_bases == 0, jnp.array(BASE_U, dtype=jnp.int8), jnp.array(BASE_C, dtype=jnp.int8))

             # Get the indices where accurate dimer repairs occur.
             accurate_dimer_indices = jnp.argwhere(accurate_dimer_repair_mask, size=num_accurate_dimer_repairs, fill_value=-1) # Get (agent_idx, base_idx)

             # Update the rna_sequence at the accurate dimer repair sites
             population_state['rna_sequence'] = jax.lax.scatter_nd_update(
                 population_state['rna_sequence'],
                 accurate_dimer_indices,
                 random_canonical_pyr,
                 scatter_dims=(0, 1) # Scatter along agent and sequence dimensions
             )


        # --- 4. Determine Stop Codons ---
        # Check for the presence of stop codons in the RNA sequence for each agent.
        # This needs to be done after mutations, as mutations can create or remove stop codons.
        # Damaged bases (8 and 99) are not canonical bases and would likely block translation if encountered by the protoribosome.
        # For stop codon check, we should only look for canonical bases.

        # Reset has_stop_codon for all agents before checking
        population_state['has_stop_codon'] = jnp.array(False, shape=NUM_AGENTS)

        # Check for each stop codon in each agent's RNA sequence
        # This part is not easily jittable due to variable inner loop length (sequence scanning)
        # For clarity and direct implementation of the concept, let's keep this for loop over agents for now.

        has_stop_codon_mask = jnp.zeros(NUM_AGENTS, dtype=bool)

        if SEQUENCE_LENGTH >= 3:
            for i in range(NUM_AGENTS):
                 rna = population_state['rna_sequence'][i]
                 is_stop = False
                 # Iterate through possible codon start positions
                 for j in range(SEQUENCE_LENGTH - 2):
                     codon = rna[j : j + 3]
                     # Check if all bases in the codon are canonical (0-3)
                     if jnp.all((codon >= 0) & (codon <= 3)):
                          # Check for stop codons
                         if jnp.array_equal(codon, STOP_CODON_UAA) or \
                            jnp.array_equal(codon, STOP_CODON_UAG) or \
                            jnp.array_equal(codon, STOP_CODON_UGA):
                             is_stop = True
                             break # Found a stop codon, no need to check further for this agent
                     # If any base in the codon is a damaged base (8 or 99), it's not a valid stop codon here
                     # and would likely cause protoribosome arrest (handled in the next step).

                 has_stop_codon_mask = has_stop_codon_mask.at[i].set(is_stop)

        population_state['has_stop_codon'] = has_stop_codon_mask


        # --- 5. Implement Protoribosome State Updates ---
        # Update protoribosome states based on current state, damage, and stop codons.
        current_proto_state = population_state['protoribosome_state']
        new_proto_state = jnp.copy(current_proto_state) # Start with current states

        # Calculate number of damaged bases per agent
        num_damaged_bases_per_agent = jnp.sum(
            (population_state['rna_sequence'] == DAMAGED_G) | (population_state['rna_sequence'] == DAMAGED_PYRIMIDINE_DIMER),
            axis=1
        )

        # Determine damage arrest threshold based on strategy
        damage_arrest_threshold_per_agent = jnp.where(
            population_state['strategy'] == 0, # Cautious strategy
            jnp.array(DAMAGE_BASE_COUNT_ARREST_THRESHOLD * CAUTIOUS_ARREST_THRESHOLD_MULTIPLIER),
            jnp.array(DAMAGE_BASE_COUNT_ARREST_THRESHOLD * RECKLESS_ARREST_THRESHOLD_MULTIPLIER) # Reckless strategy
        )


        # Transition rules:
        # ACTIVE (0) -> ARRESTED (1): if has_stop_codon is True OR num_damaged_bases_per_agent > strategy-dependent threshold
        # ARRESTED (1) -> INACTIVE (2) : if protoribosome_arrest_timer exceeds ARREST_DURATION_STEPS
        # ARRESTED (1) -> ACTIVE (0): if damaged bases are repaired AND stop codon removed (simplified: if no stop codon AND num_damaged_bases_per_agent <= strategy-dependent threshold)
        # INACTIVE (2) -> ACTIVE (0): (Optional, re-initiation) after a period of inactivity (e.g., if timer exceeds a threshold) AND conditions for arrest are no longer met

        # Update protoribosome arrest timer for agents currently in ARRESTED state
        arrested_mask = (current_proto_state == PROTO_STATE_ARRESTED)
        protoribosome_arrest_timer = jnp.where(
            arrested_mask,
            population_state['protoribosome_arrest_timer'] + 1, # Use protoribosome_arrest_timer from population_state
            jnp.array(0, dtype=jnp.int32) # Timer resets for other states
        )

        # Transition from ACTIVE to ARRESTED
        active_to_arrested_mask = (current_proto_state == PROTO_STATE_ACTIVE) & \
                                   (population_state['has_stop_codon'] | (num_damaged_bases_per_agent > damage_arrest_threshold_per_agent))

        new_proto_state = jnp.where(
            active_to_arrested_mask,
            jnp.array(PROTO_STATE_ARRESTED, dtype=jnp.int8), # Ensure dtype matches population_state
            new_proto_state
        )


        # Transition from ARRESTED to INACTIVE
        # This transition is based on the arrest timer
        arrested_to_inactive_mask = (current_proto_state == PROTO_STATE_ARRESTED) & \
                                      (protoribosome_arrest_timer >= jnp.array(ARREST_DURATION_STEPS))

        new_proto_state = jnp.where(
            arrested_to_inactive_mask,
            jnp.array(PROTO_STATE_INACTIVE, dtype=jnp.int8), # Ensure dtype matches population_state
            new_proto_state
        )

        # Transition from ARRESTED back to ACTIVE (Recovery)
        # Conditions: No stop codon AND damage is below the strategy-dependent threshold
        arrested_to_active_mask = (current_proto_state == PROTO_STATE_ARRESTED) & \
                                  (~population_state['has_stop_codon'] & (num_damaged_bases_per_agent <= damage_arrest_threshold_per_agent))

        new_proto_state = jnp.where(
            arrested_to_active_mask,
            jnp.array(PROTO_STATE_ACTIVE, dtype=jnp.int8),
            new_proto_state
        )
        # Reset timer for agents transitioning from ARRESTED to ACTIVE
        protoribosome_arrest_timer = jnp.where(
            arrested_to_active_mask,
            jnp.array(0, dtype=jnp.int32),
            protoribosome_arrest_timer
        )


        # Transition from INACTIVE back to ACTIVE (Re-initiation)
        # Let's add a simple re-initiation probability for inactive protoribosomes
        # if conditions for arrest are not met.
        inactive_to_active_mask_conditions = (current_proto_state == PROTO_STATE_INACTIVE) & \
                                             (~population_state['has_stop_codon'] & (num_damaged_bases_per_agent <= damage_arrest_threshold_per_agent))

        # Add a re-initiation probability
        key, reinit_roll_key = jax.random.split(key)
        reinit_roll = jax.random.uniform(reinit_roll_key, shape=inactive_to_active_mask_conditions.shape)
        inactive_to_active_mask = inactive_to_active_mask_conditions & (reinit_roll < REINITIATION_RATE)


        new_proto_state = jnp.where(
            inactive_to_active_mask,
            jnp.array(PROTO_STATE_ACTIVE, dtype=jnp.int8),
            new_proto_state
        )
        # Reset timer for agents transitioning from INACTIVE to ACTIVE (timer is already 0)


        # Update the protoribosome_state in the population_state
        population_state['protoribosome_state'] = new_proto_state
        # Update the arrest timer in the population_state
        population_state['protoribosome_arrest_timer'] = protoribosome_arrest_timer


        # --- 7. Implement Autocatalysis (Success/Failure and Population Dynamics) ---
        # Determine which agents are capable of successful autocatalysis (reproduction)
        # Only agents with ACTIVE protoribosomes can potentially reproduce
        can_reproduce_mask = (population_state['protoribosome_state'] == PROTO_STATE_ACTIVE)
        key, success_roll_key = jax.random.split(key)
        success_roll = jax.random.uniform(success_roll_key, shape=can_reproduce_mask.shape)
        successfully_reproducing_agents_mask = can_reproduce_mask & (success_roll < AUTOCATALYSIS_SUCCESS_RATE_ACTIVE)

        num_successfully_reproducing_agents = jnp.sum(successfully_reproducing_agents_mask)

        # Update autocatalysis failure timer: increment for agents that did NOT reproduce, reset for those that did.
        autocatalysis_failure_timer = jnp.where(
            successfully_reproducing_agents_mask,
            jnp.array(0, dtype=jnp.int32), # Reset timer on successful reproduction
            population_state['autocatalysis_failure_timer'] + 1 # Increment timer on failure to reproduce
        )
        population_state['autocatalysis_failure_timer'] = autocatalysis_failure_timer


        # --- 8. Implement Foraging/Removal ---
        # Agents can be removed from the population (foraged) if they fail autocatalysis
        # consistently for a certain period, or based on other criteria.
        # Use the autocatalysis_failure_timer to determine foraging risk.

        # Identify agents at risk of foraging (failed for a certain duration)
        foraging_risk_mask = (population_state['autocatalysis_failure_timer'] >= jnp.array(FAILURE_DURATION_FORAGING_THRESHOLD))

        # Determine which agents are foraged
        key, foraging_roll_key = jax.random.split(key)
        foraging_roll = jax.random.uniform(foraging_roll_key, shape=foraging_risk_mask.shape)
        foraged_agents_mask = foraging_risk_mask & (foraging_roll < FORAGING_RATE_FAILED)

        num_foraged_agents = jnp.sum(foraged_agents_mask)

        # Implement Population Dynamics: Replace foraged agents with offspring from successfully reproducing agents
        # This maintains a constant population size (NUM_AGENTS).
        num_replacements_needed = num_foraged_agents
        num_potential_offspring = num_successfully_reproducing_agents

        # If there are reproducing agents and replacements needed, select reproducing agents
        if num_potential_offspring > 0 and num_replacements_needed > 0:
             key, offspring_selection_key = jax.random.split(key)
             # Select indices of successfully reproducing agents (potential parents)
             reproducing_indices = jnp.where(successfully_reproducing_agents_mask)[0]

             # Randomly select parents to produce offspring to replace foraged agents
             # Use replacement=True if num_replacements_needed > num_potential_offspring
             replace_with_replacement = num_replacements_needed > num_potential_offspring
             parent_indices_for_offspring = jax.random.choice(
                 offspring_selection_key,
                 reproducing_indices,
                 shape=(num_replacements_needed,),
                 replace=replace_with_replacement
             )

             # Get the state of the selected parent agents
             parent_rna_sequences = population_state['rna_sequence'][parent_indices_for_offspring]
             parent_strategies = population_state['strategy'][parent_indices_for_offspring]

             # Create offspring state (e.g., copy parent's RNA and strategy, reset damage/protoribosome state, reset timers)
             offspring_rna_sequences = jnp.copy(parent_rna_sequences) # Offspring inherit RNA
             offspring_strategies = jnp.copy(parent_strategies) # Offspring inherit strategy
             offspring_rna_damage = jnp.zeros_like(offspring_rna_sequences, dtype=jnp.float32) # Offspring start with no general damage
             # Important: Stop codons need to be re-checked for offspring based on inherited sequence in the next step
             offspring_has_stop_codon = jnp.zeros(num_replacements_needed, dtype=bool)
             offspring_protoribosome_state = jnp.full(num_replacements_needed, PROTO_STATE_ACTIVE, dtype=jnp.int8) # Offspring protoribosomes are active
             offspring_protoribosome_arrest_timer = jnp.zeros(num_replacements_needed, dtype=jnp.int32) # Offspring timer is 0
             offspring_autocatalysis_failure_timer = jnp.zeros(num_replacements_needed, dtype=jnp.int32) # Offspring timer is 0


             # Select indices of foraged agents to be replaced
             foraged_indices = jnp.where(foraged_agents_mask)[0]

             # Update the state of the foraged agents with the offspring state
             # Use jax.lax.scatter_nd_update to update multiple indices at once
             population_state['strategy'] = jax.lax.scatter_nd_update(
                 population_state['strategy'],
                 foraged_indices[:, None], # Indices need to be shape (N, 1) for scatter
                 offspring_strategies,
                 scatter_dims=(0,) # Scatter along the first dimension (agents)
             )
             population_state['rna_damage'] = jax.lax.scatter_nd_update(
                 population_state['rna_damage'],
                 foraged_indices[:, None, None], # Indices need to be shape (N, 1, 1) for scatter into 2D array
                 offspring_rna_damage,
                 scatter_dims=(0,)
             )
             population_state['has_stop_codon'] = jax.lax.scatter_nd_update(
                 population_state['has_stop_codon'],
                 foraged_indices[:, None],
                 offspring_has_stop_codon,
                 scatter_dims=(0,)
             )
             population_state['rna_sequence'] = jax.lax.scatter_nd_update(
                 population_state['rna_sequence'],
                 foraged_indices[:, None, None], # Indices need to be shape (N, 1, 1) for scatter into 2D array
                 offspring_rna_sequences,
                 scatter_dims=(0,)
             )
             population_state['protoribosome_state'] = jax.lax.scatter_nd_update(
                 population_state['protoribosome_state'],
                 foraged_indices[:, None],
                 offspring_protoribosome_state,
                 scatter_dims=(0,)
             )
             population_state['protoribosome_arrest_timer'] = jax.lax.scatter_nd_update(
                 population_state['protoribosome_arrest_timer'],
                 foraged_indices[:, None],
                 offspring_protoribosome_arrest_timer,
                 scatter_dims=(0,)
             )
             population_state['autocatalysis_failure_timer'] = jax.lax.scatter_nd_update(
                 population_state['autocatalysis_failure_timer'],
                 foraged_indices[:, None],
                 offspring_autocatalysis_failure_timer,
                 scatter_dims=(0,)
             )

        # If no reproducing agents but foraging occurs, these spots are effectively removed (status=0 implicitly, or just replaced with empty/dummy state)
        # In this replacement model, if num_potential_offspring is 0, num_replacements_needed will also be 0 as no offspring can be created.
        # If num_successfully_reproducing_agents < num_foraged_agents, we replace as many as possible from reproducing agents.
        # The remaining foraged spots are effectively empty until future steps produce more offspring than failures.


        # --- Placeholder for other simulation step logic ---
        # (All core simulation step logic is now implemented)


        # Placeholder print statement
        if step % 1000 == 0 or step == start_step or step == NUM_STEPS:
            # Calculate average damaged bases per agent for printing
            num_damaged_bases_per_agent = jnp.sum(
                (population_state['rna_sequence'] == DAMAGED_G) | (population_state['rna_sequence'] == DAMAGED_PYRIMIDINE_DIMER),
                axis=1
            )
            avg_damaged_bases_for_print = jnp.mean(num_damaged_bases_per_agent) if num_damaged_bases_per_agent.shape[0] > 0 else 0.0

            num_agents_with_stop_codon = jnp.sum(population_state['has_stop_codon']) # Count agents with stop codons
            # Count protoribosome states for all agents
            active_protos_count = jnp.sum(population_state['protoribosome_state'] == PROTO_STATE_ACTIVE)
            arrested_protos_count = jnp.sum(population_state['protoribosome_state'] == PROTO_STATE_ARRESTED)
            inactive_protos_count = jnp.sum(population_state['protoribosome_state'] == PROTO_STATE_INACTIVE)

            # Count agents by strategy
            cautious_count = jnp.sum(population_state['strategy'] == 0)
            reckless_count = jnp.sum(population_state['strategy'] == 1)

            # Total agents remain constant at NUM_AGENTS in this replacement model


            print(f"Simulating step {step}/{NUM_STEPS}, UV: {current_uv_radiation:.2f}, Temp: {current_temperature:.0f}C, Avg Damaged Bases: {avg_damaged_bases_for_print:.2f}, Error-Prone Mutations: {num_error_prone_mutations}, Agents with Stop Codon: {num_agents_with_stop_codon}")
            print(f"  Protoribosome States: Active={active_protos_count}, Arrested={arrested_protos_count}, Inactive={inactive_protos_count}")
            print(f"  Agents by Strategy: Cautious={cautious_count}, Reckless={reckless_count}")
            print(f"  Autocatalysis: Successful={num_successfully_reproducing_agents}, Failed (this step): {NUM_AGENTS - num_successfully_reproducing_agents}, Foraged: {num_foraged_agents}")


        # Add checkpointing logic
        # Only save if population_state is defined
        if population_state is not None and step > 0 and (step % CHECKPOINT_INTERVAL == 0 or step == NUM_STEPS):
            if step == NUM_STEPS:
                 checkpoint_filename = os.path.join(CHECKPOINT_DIR, f'final_state_step_{step}.pkl')
            else:
                 checkpoint_filename = os.path.join(CHECKPOINT_DIR, f'checkpoint_step_{step}.pkl')

            try:
                with open(checkpoint_filename, 'wb') as f:
                    pickle.dump(population_state, f)
                print(f"Checkpoint saved at step {step} to {checkpoint_filename}")
            except Exception as e:
                print(f"Error saving checkpoint at step {step}: {e}")

else:
    print("Simulation loop skipped because population_state is not defined.")


print("Simulation loop finished.")
```
